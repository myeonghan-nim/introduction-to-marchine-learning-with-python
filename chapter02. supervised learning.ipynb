## 2.1 분류와 회귀

- 지도 학습은 크게 **분류**와 **회귀**로 나눕니다.

    - 분류는 미리 정의된 여러 **클래스 레이블** 중 하나를 예측하는 문제입니다.
    
    > 앞서서 살펴본 붓꽃 예측하기가 분류 문제 중 하나입니다.
    
- 분류는 두 개의 클래스로 분류하는 **이진 분류**와 셋 이상 클래스로 분류하는 **다중 분류**로 나뉩니다.

    - 이진 분류는 답이 예 혹은 아니오로 나오도록 만드는 것이라고 생각하면 됩니다.

    > 예를 들어 특정 이메일이 스팸 메일인지 아닌지 판별하는 것이 이진 분류 문제 중 하나입니다.

#### Note

- 이진 분류에서 한 클래스를 **양성** 클래스, 다른 하나를 **음성** 클래스라고 합니다.

    - 양성 클래스라고 해서 좋은 값이나 장점을 말하는 것은 아니며 학습하고자 하는 대상을 의미합니다.
    
    > 예를 들면 스팸 메일의 경우 양성이 스팸 클래스 입니다.

- 두 클래스 중 어떤 것을 양성으로 정할지는 주관적인 문제로 분야에 따라 달라집니다.

- 분류 문제 중 다른 하나는 다중 분류로 여러 개의 정답 중 하나를 찾는 것을 의미합니다.

    - 즉, 다중 분류는 답이 여러 개가 존재하는 것을 의미합니다.

    > 앞서서 본 붓꽃 분류 문제나 해당 글이 어떤 언어로 작성된 글인지 맞추는 문제 등이 다중 분류 문제입니다.

- 회귀는 연속적인 숫자 또는 프로그래밍 용어로 **부동소수점수**(수학 용어로 **실수**)를 예측하는 것을 말합니다.

    - 예를 들어 어떤 사람의 교육 수준, 나이, 주거지를 바탕으로 연간 소득을 예측 하는 것이 회귀 문제 중 하나입니다.
    
    - 그 외에도 옥수수 농장의 전년도 수확량과 날씨, 고용 인원 수 등을 바탕으로 올해 수확량을 예측하는 것도 한 문제입니다.

- 출력 값에 연속성이 있는지 질문해보면 회귀와 분류 문제를 쉽게 구분할 수 있습니다.

    - 예상 출력 값 사이에 연속성이 있다면 회귀 문제입니다.
    
        - 예를 들어 연소득을 예측하는 문제는 예측한 정답에서 1 ~ 2원 차이가 났다고 해서 큰 문제가 일어나지 않습니다.
        
        - 반면에 언어를 맞추는 문제는 어느 정도란 것이 없습니다. 즉, 언어는 어느 한 언어가 아니라면 다른 언어입니다.

## 2.2 일반화, 과대적합, 과소적합

- 지도 학습에서 훈련 데이터로 학습한 모델이 훈련 데이터와 특성이 같다면 처음 보는 새로운 데이터도 정확히 예측할 거라 기대합니다.

- 모델이 처음 보는 데이터에 대해 정확하게 예측할 수 있으면 이를 훈련 세트에서 테스트 세트로 **일반화** 되었다고 합니다.

    - 그래서 모델을 만들 때는 가능한 한 정확하게 일반화되도록 해야 합니다.

- 보통 훈련 세트에 대해 정확히 예측하도록 모델을 구축합니다. 이 때 훈련 세트와 테스트 세트가 매우 비슷하다면 그 모델이 테스트 세트에서도 정확히 예측하리라 기대할 수 있습니다.

- 그러나 항상 그런 것만은 아닙니다. 예로 아주 복잡한 모델을 만들면 훈련 세트에만 정확한 모델이 되어버릴 수 있습니다.

- 알고리즘이 새로운 데이터도 잘 처리하는지 측정하는 방법은 테스트 세트로 평가해보는 것밖에 없습니다.

    - 그러나 직관적으로 보더라도 간단한 모델이 새로운 데이터에 더 일반화될 것이라고 예상할 수 있습니다.

- 이에 비추어 가진 정보를 모두 사용해 너무 복잡한 모델을 만드는 것을 **과대적합**이라고 합니다.

    - 과대 적합은 모델이 훈련 세트의 각 샘플에 너무 가깝게 맞춰져서 새로운 데이터에 일반화되기 어려울 때 일어납니다.

- 반대로 너무 간단한 모델은 데이터의 면면과 다양성을 잡아내지 못할 것이며 훈련 세트와도 맞지 않을 것입니다. 이를 **과소적합**이라고 합니다.

- 즉, 모델이 복잡할수록 훈련 데이터에 대해 더 정확히 예측할 수 있으나 너무 복잡해지면 훈련 세트의 각 데이터 포인트에 너무 민감해져 새로운 데이터에 잘 일반화되지 못합니다.

> 즉, 우리가 만들어야하는 모델은 일반화 성능이 최대가 되는 최적점에 있는 모델입니다.

### 2.2.1 모델 복잡도와 데이터셋 크기의 관계

- 모델 복잡도는 훈련 데이터셋에 담긴 입력 데이터의 다양성과 관련이 깊습니다.

    - 데이터셋에 다양한 데이터 포인트가 많을수록 과대적합 없이 더 복잡한 모델을 만들 수 있습니다.
    
    - 보통 데이터 포인트를 더 많이 모으는 것이 다양성을 키워주므로 큰 데이터셋은 더 복잡한 모델을 만들 수 있게 해줍니다.
    
    - 다만, 같은 데이터 포인트를 중복하거나 매우 비슷한 데이터를 모으는 것은 도움이 되지 않습니다.

- 데이터를 더 많이 수집하고 적절하게 더 복잡한 모델을 만들면 지도 학습 문제에서 놀라운 결과를 얻을 수 있습니다.

    > 이 책에서는 고정 크기 데이터셋을 사용할 것입니다.
    >
    > 이는 모델을 변경하거나 조정하는 것보다 이득일 수 있으나 실제 환경에서는 데이터를 얼마나 많이 모을지 정해야 합니다.

### 2.3.2 k-최근접 이웃

- k-NN 알고리즘은 가장 간단한 머신러닝 알고리즘입니다. 훈련 데이터셋을 그냥 저장하는 것이 모델을 만드는 과정의 전부입니다.

- 새로운 데이터 포인트에 대해 예측할 땐 알고리즘이 가장 가까운 데이터 포인트, 즉 '최근접 이웃'을 찾습니다.

#### k-최근접 이웃 분류

- 가장 간단한 k-NN 알고리즘은 가장 가까운 훈련 데이터 포인트 하나를 최근접 이웃으로 찾아 예측에 사용합니다.

- 단순히 이 훈련 데이터 포인트의 출력이 예측되면 다음과 같이 분류할 수 있습니다.

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

- 위 그래프는 데이터 포인트 3개(별 모양)에 대해 각각 가장 가까운 훈련 데이터 포인트 1개를 연결한 것입니다.

    - 이 반식이 1-최근접 이웃 알고리즘의 예측으로 데이터 포인트의 레이블이 됩니다.

- 반면 가장 가까운 이웃을 하나가 아닌 임의의 k개를 선택할 수도 있습니다. 바로 이 경우가 k-최근접 이웃 알고리즘입니다.

    - 둘 이상의 이웃을 선택할 때는 레이블을 정하기 위한 투표를 합니다.
    
    - 즉, 테스트 포인트 하나에 대해 각 클래스에 속한 이웃이 몇 개인지 셉니다. 그리고 이웃이 더 많은 클래스를 레이블로 지정합니다.
    
    - 다시 말해 k-최근접 이웃 중 다수의 클래스가 레이블이 됩니다. 다음은 세 개의 최근접 이웃을 사용하는 예입니다.

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

- 여기서도 예측은 연결된 데이터 포인트의 색으로 나타납니다. 이를 살펴보면 이웃을 하나만 선택할 때와 결과가 달라진 데이터를 알 수 있습니다.

- 이 예시는 이진 분류 문제이지만 클래스가 다수인 데이터셋도 같은 방법을 적용할 수 있습니다.

    - 클래스가 여러 개 일때도 각 클래스에 속한 이웃이 몇 개인지를 헤아려 가장 많은 클래스를 예측값으로 사용합니다.

> 이제 scikit-learn을 사용해 k-최근접 이웃 알고리즘을 어떻게 적용하는지 살펴볼 것입니다.

```python
from sklearn.model_selection import train_test_split

# 일반화 성능을 평가하기 위해 데이터를 훈련 세트와 테스트 세트로 나눕니다.
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# k-NN 클래스를 import하고 하나의 객체로 만듭니다. 이 때 이웃의 수 같은 매개변수를 지정합니다.
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3)

# 훈련 세트를 사용해 분류 모델을 학습시킵니다.
# k-NN의 경우 에측할 때 이웃을 찾을 수 있도록 데이터를 저장합니다.
clf.fit(X_train, y_train)

# 테스트 세트를 predict를 사용해 예측합니다.
# k-NN은 각 데이터 포인트에 대해 훈련 세트에서 가장 가까운 이웃을 계산한 다음 가장 많은 클래스를 찾습니다.
print('테스트 세트 예측:', clf.predict(X_test))

# 모델이 얼마나 잘 일반화되었는지 score를 사용해 평가합니다.
print(f'테스트 세트 정확도: {clf.score(X_test, y_test):.2f}')
```

#### KNeighborsClassifier 분석

- 2차원 데이터셋이므로 가능한 모든 테스트 포인트의 예측을 xy 평면에 그릴 수 있습니다.

    - 각 데이터 포인트가 속한 클래스에 따라 평면에 색을 칠하여 알고리즘이 지정한 클래스로 나뉘는 **결정 경계**를 볼 수 있습니다.
    
> 다음 코드는 이웃이 1, 3, 9개일 때 결정 경계를 보여줍니다.

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    # fit 메서드는 self 오브젝트를 리턴하므로 객체 생성과 fit을 동시에 할 수 있습니다.
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    
    # 그래프를 그립니다.
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    
    # 각 그래프의 속성을 부여합니다.
    ax.set_title(f'{n_neighbors} neighbors')
    ax.set_xlabel('attribute 0')
    ax.set_ylabel('attribute 1')

axes[0].legend(loc=3)
```

- 위 그래프를 보면 이웃이 적으면 결정 경계가 훈련 데이터에 가깝게 따라가고 많으면 경계가 부드러워짐을 알 수 있습니다.

- 다시 말해 이웃을 적게 사용하면 모델의 복잡도가 높아지고 이웃을 많이 사용하면 복잡도가 낮아집니다.

    - 극단적으로 이웃의 수를 훈련 데이터 전체로 지정하면 모든 테스트 포인트가 같은 이웃을 가져 테스트 포인트에 대한 예측이 같은 값이 됩니다.
    
    - 즉, 훈련 세트에서 가장 많은 데이터 포인트를 가진 클래스가 예측값이 됩니다.

- 이어서 모델의 복잡도와 일반화 사이의 관계를 실제 데이터인 유방암 데이터셋을 사용해 확인하겠습니다.

```python
from sklearn.datasets import load_breast_cancer

# 데이터 객체를 저장합니다.
cancer = load_breast_cancer()

# 데이터셋을 훈련 세트와 테스트 세트로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66)

# 정확도를 평가합니다.
training_accuracy, test_accuracy = [], []
neighbors_setting = range(1, 11)
for n_neighbors in neighbors_setting:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

# 그래프를 그립니다.
plt.plot(neighbors_setting, training_accuracy, label='training accuracy')
plt.plot(neighbors_setting, test_accuracy, label='test accuracy')
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.legend()
```

- 위 그래프에 따르면 n_neighbors 수에 따를 훈련 세트와 테스트 세트의 정확도를 알 수 있습니다.

- 실제 그래프는 이렇게 매끈하게 나오지 않지만 여기서 과대적합과 과소적합의 특징을 볼 수 있습니다.

    - 최근접 이웃의 수가 하나일 때 훈련 데이터에 대한 예측이 완벽하나 이웃의 수가 늘어나면 모델은 단순해져 훈련 세트의 정확도가 줄어듭니다.
    
    - 반면에 이웃을 하나만 사용한 테스트 세트의 정확도는 이웃을 많이 사용할 때보다 낮습니다.
    
    - 이는 모델이 너무 복잡함을 설명하며 반대로 이웃이 많을 땐 모델이 너무 단순해 정확도가 나빠집니다.
    
    > 즉, 정확도가 가장 좋을 때는 중간 정도인 6개를 사용한 경우입니다.

#### k-최근접 이웃 회귀

- k-최근접 이웃 알고리즘은 회귀 분석에도 사용됩니다. 이번에는 wave 데이터셋을 이용해 이를 살펴보겠습니다.

    - x축에 세 개의 테스트 데이터를 흐린 별 모양으로 표시했습니다.
    
    - 여기서 최근접 이웃을 한 개만 사용할 때 예측값은 가장 가까운 이웃의 타깃값입니다.

```python
mglearn.plots.plot_knn_regression(n_neighbors=1)
```

- 한 편 이웃을 둘 이상 사용하여 회귀 분석을 진행할 수 있으며 여러 개의 최근접 이웃을 사용하는 경우 이웃간의 평균으로 계산됩니다.

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
```

- scikit-learn에서 회귀를 사용한 k-최근접 이웃 알고리즘은 KNeighborsRegressor로 구현되어 있습니다.

> 사용법은 KNeighborsClassifier와 비슷합니다.

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

# wave 데이터셋을 훈련 세트와 테스트 세트로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# 이웃의 수를 3으로 해 모델 객체를 만들고 학습과 예측을 진행합니다.
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)  # 학습

print('테스트 세트 예측:\n', reg.predict(X_test))
```

- 반면 k-최근접 이웃 회귀 알고리즘도 score 메서드를 사용해 모델을 평가할 수 있습니다.

    - score는 회귀의 경우 R<sup>2</sup> 값을 통해 계산합니다.
    
    - 결정 계수라고도 불리는 이 값은 0과 1사이의 값을 가지며 0 일수록 y_train의 평균으로만 예측하며 1 일수록 예측이 완벽합니다.
    
    - R<sup>2</sup>는 음수가 될 수도 있으며 이 경우 예측과 타깃이 서로 상반되는 결과를 가져오는 경우를 의미합니다.

```python
print(f'테스트 세트 스코어: {reg.score(X_test, y_test):.2f}')
```

#### KNeighborsRegressor 분석

- 이 데이터셋에 대해 가능한 모든 특성 값을 만들어 예측해 보겠습니다. 이를 위해 x축을 따라 많은 포인트를 생성해 테스트 데이터셋을 만듭니다.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# -3부터 3 사이에 1000개의 데이터 포인트를 만듭니다.
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
for n_neighbors, ax in zip([1, 3, 9], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)  # 학습
    reg.fit(X_train, y_train)
    
    ax.plot(line, reg.predict(line))  # 예측
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(f'{n_neighbors} train score: {reg.score(X_train, y_train):.2f} / test score: {reg.score(X_test, y_test):.2f}')  # 그래프
    ax.set_xlabel('attributes')
    ax.set_ylabel('target')

axes[0].legend(['model predict', 'train data/target', 'test data/target'], loc='best')
```

- 이처럼 이웃이 적은 경우 훈련 세트의 각 데이터 포인트가 예측에 주는 영향이 커, 예측값이 훈련 데이터 포인트를 모두 지나갑니다.

- 이는 매우 불안정한 예측을 만들어내는 반면에 이웃이 많아지만 훈련 데이터에는 잘 안맞을 수 있으나 더 안정된 예측을 얻어냅니다.

#### 장단점과 매개변수

- 일반적으로 KNeighbors 분류기에 중요한 매개변수는 두 개로 데이터 포인트 사이의 거리를 재는 방법과 이웃의 수입니다.

    - 실제로 이웃의 수가 3 ~ 5개로 적을 때 잘 작동하지만, 이 매개변수는 잘 조절해야 합니다.
    
    - 반면 거리를 재는 방법은 일반적인 환경에서 잘 동작하는 유클라디안 거리 방식을 사용합니다.

- k-NN의 장점은 이해하기 쉬운 모델이라는 점입니다.

    - 많이 조정하지 않아도 자주 좋은 성능을 발휘해 더 복잡한 알고리즘을 적용해보기 전에 시도해볼 수 있는 좋은 시작점입니다.
    
    - 보통 최근접 이웃 모델은 매우 빠르게 만들 수 있으나 훈련 세트가 커지면 예측이 느려집니다.
    
    - 따라서 k-NN 알고리즘을 사용할 땐 데이터를 전처리하는 과정이 중요합니다.
    
- 반면에 많은 특성을 가진 데이터셋에는 잘 동작하지 않으며 특성 값 대부분이 거의 없는 데이터셋에 특히 잘 작동하지 않습니다.

- k-최근접 이웃 알고리즘의 이해하기 쉽지만 예측이 느리고 많은 특성을 처리하는 능력이 부족함을 보완한 모델이 바로 선형 모델입니다.

> 따라서 현업에서 k-최근접 이웃 알고리즘은 잘 쓰이지 않습니다.

### 2.3.3 선형 모델

- 선형 모델은 100여 년 전에 개발되었고 지난 몇 십년간 폭넓게 연구되어 현재도 널리 쓰이는 모델입니다.

- 선형 모델은 입력 특성에 대한 **선형 함수**를 만들어 예측을 수행합니다.

#### 회귀의 선형 모델

- 회귀의 경우 선형 모델을 위한 일반화된 예측 함수(`y = w[0] * x[0] + w[1] * x[1] + ... + w[p] + x[p] + b`)를 사용합니다.

    - 위 식에서 x는 하나의 데이터 포인트에 대한 특성이며 w와 b는 모델이 학습할 파라미터입니다. 그리고 y는 모델이 예측한 값입니다.
    
    > 만일 특성이 하나인 데이터셋이라면 위의 식은 `y = w[0] * x[0] + b`로 간결해집니다.
    
    - 수학 시간에 배운 직선의 방정식처럼 w는 x의 기울기이고 b는 y축과 만나는 절편입니다.
    
    - 여기에 특성이 많아지면 w는 각 특성에 해당하는 기울기ㄴ를 모두 가지므로 예측값은 입력 특성에 w의 각 가중치(음수 포함)를 곱해 더한 가중치의 합입니다.

- 이제 1차원 wave 데이터셋으로 파라미터 w와 b를 학습시킨 결과를 살펴보겠습니다.

```python
mglearn.plots.plot_linear_regression_wave()
```

> 위 그래프는 이해하기 쉽도록 그래프의 중앙을 가로질러 x, y축을 그렸습니다.

- 회귀를 위한 선형 모델은 특성이 하나일 때 직선, 두 개일 때 평면이 되며 그 이상에서는 초평면이 되는 회귀 모델의 특징을 가지고 있습니다.

- 이 직선과 KNeighborsRegressor를 사용하여 만든 선을 비교하면 직선을 사용한 예측의 제약이 더 많아 보입니다.

    - 즉, 데이터의 상세 정보를 모두 잃어버린 것처럼 보이며 이는 어느 정도 사실입니다.

- 하지만, 타깃 y과 특성들의 선형 조합임을 생각하면 매우 과한 가정으로 이는 1차원 데이터만 놓고 봐서 생긴 편견일 수 있습니다.

- 특성이 많은 데이터셋이라면 선형 모델은 매우 훌륭한 성능을 낼 수 있습니다.

    - 특히, 훈련 데이터보다 특성이 더 많은 경우엔 훈련 세트에 대해 어떤 타깃 y도 선형 함수로 모델링할 수 있습니다.

- 회귀를 위한 선형 모델은 매우 다양한 데, 이 모델들은 훈련 데이터에서 파라미터 w와 b를 학습하는 방법과 모델의 복잡도를 제어하는 방법에서 차이가 발생합니다. 그럼 하나씩 해당 모델들을 배워보겠습니다.

#### 선형 회귀(최소제곱법)

- **선형 회귀** 혹은 **최소제곱법**은 가장 간단하고 오래된 회귀용 선현 알고리즘입니다.

    - 선형 회귀는 예측과 훈련 세트에 있는 타깃 y 사이의 **평균제곱오차**를 최소화하는 파라미터 w와 b를 찾습니다.
    
    > 평균제곱오차란 예측값과 타깃값의 차이를 제곱하여 더한 후 샘플의 개수로 나눈 것입니다.
    
- 선형회귀는 매개변수가 없는 것이 장점이지만 모델으리 복잡도를 제어할 방법도 없습니다.

> 다음은 선형 회귀를 만드는 코드입니다.

```python
from sklearn.linear_model import LinearRegression

X, y = mglearn.datasets.make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 모델을 학습합니다.
lr = LinearRegression().fit(X_train, y_train)
```

```python
# 기울기 파라미터 w는 가중치 또는 계수라고 부르며 lr 객체의 coef_ 속성에 저장되어 있습니다.
print('lr.coef_:', lr.coef_)
# 편향 파라미터 b는 절편이라고도 불리며 lr 객체의 intercept_ 속성에 저장되어 있습니다.
print('lr.intercept_:', lr.intercept_)
```

#### Note

- coef_, intercept_와 같은 속성의 뒤에 `_`가 붙은 이유는 scikit-learn의 특징으로 훈련 데이터에서 유도된 속성을 의미합니다.

- 이는 사용자가 지정한 매개변수와 구분하기 위해서 입니다.

- intercept_ 속성은 항상 실수 하나지만, coef_ 속성은 각 입력 특성에 일대일대응하는 NumPy 배열입니다.

    - wave 데이터셋은 입력 특성이 하나뿐이므로 lr.coef_도 원소를 하나만 가지고 있습니다.

> 다음은 훈련 세트와 테스트 세트의 성능 확인입니다.

```python
print(f'훈련 세트 점수: {lr.score(X_train, y_train):.2f}')
print(f'테스트 세트 점수: {lr.score(X_test, y_test):.2f}')
```

- 결과가 0.66으로 그다지 좋은 결과가 아니지만 훈련 세트와 테스트 세트의 점수가 비슷함을 알 수 있습니다.

    - 이는 과대적합 혹은 과소적합인 상태임을 의미합니다.

- 1차원 데이터셋은 모델이 단순하므로 과대적합을 걱정할 필요가 없으나 고차원 데이터셋은 선형 모델의 성능이 매우 높아져 과대적합될 가능성이 높습니다.

    - 이와 비슷하게 LinearRegression 모델이 보스턴 주택가격 데이터셋에 어떻게 동작하는지 알아보겠습니다.
    
    - 이 데이터셋은 샘플이 506개이며 유도된 특성을 합쳐 104개의 특성을 가지고 있습니다.

```python
# 데이터를 불러들입니다.
X, y = mglearn.datasets.load_extended_boston()

# 데이터를 나누고 모델을 만듭니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
lr = LinearRegression().fit(X_train, y_train)
```

- 결과를 확인하면 훈련 세트의 점수가 매우 높아 예측이 정확하나 테스트 세트의 점수가 낮은 걸 알 수 있습니다.

```python
print(f'훈련 세트 점수: {lr.score(X_train, y_train):.2f}')
print(f'테스트 세트 점수: {lr.score(X_test, y_test):.2f}')
```

- 이는 모델이 과대적합되었다는 신호로 이를 해결하기 위해서 모델의 복잡도를 제어할 수 있는 모델을 사용해야 합니다.

- 따라서 기본 선형 회귀 방식 대신 가장 널리 쓰이는 모델은 **리지 회귀** 모델입니다.

#### 리지 회귀

- **리지**도 회귀를 위한 선형 모델로 최소적합법에서 사용한 것과 같은 예측 함수를 사용합니다.

    - 하지만 리지 회귀에서 가중치 선택은 훈련 데이터를 잘 예측하기 위해서만이 아니라 추가 제약 조건을 만족시키려는 목적도 있습니다.
    
    - 이는 가중치의 절댓값을 가능한 작게 만드는 것으로 직관적으로 이는 모든 특성이 출력에 주는 영향을 최소한으로 만듭니다.
    
    - 바로 이런 제약을 **규제**라고 하며 규제는 과대적합이 되지 않도록 모델을 강제로 제한하는 것을 말합니다.

- 리지 회귀에서는 **L2 규제**를 사용합니다.

- 리지 회귀는 `sklearn.linear_model.Ridge`에 구현되어 있습니다. 이를 사용해 확장된 보스턴 주택가격 데이터셋을 계산해보겠습니다.

```python
from sklearn.linear_model import Ridge

ridge = Ridge().fit(X_train, y_train)
print(f'훈련 세트 점수: {ridge.score(X_train, y_train):.2f}')
print(f'테스트 세트 점수: {ridge.score(X_test, y_test):.2f}')
```

- 결과를 보면 훈련 세트의 점수는 LinearRegression 보다 낮지만 테스트 세트 점수는 더 높습니다.

- 이는 선형 회귀는 이 데이터셋에 과대적합하지만 Ridge는 덜 자유로운 모델이기 때문입니다.

    - 모델의 복잡도가 낮아지면 훈련 세트의 성능은 나빠지마 더 일반화된 모델이 됩니다.
    
    - 우리에게 중요한 건 테스트 세트에 대한 성능이므로 해당 결과가 더 높은 Ridge를 선택해야 합니다.

- Ridge는 모델을 단순하게 해 훈련 세트에 대한 성능 차이를 절충하는 방법을 제공하는데 이 변수가 바로 `alpha`입니다.

    - 사용자는 alpha로 훈련 세트의 성능 대비 모델을 얼마나 단순화할 지 지정할 수 있습니다.(기본값은 1.0입니다.)
    
    - 하지만 이 값이 최적일 수는 없는데 최적 alpha는 데이터셋에 따라 다를 수 있기 때문입니다.

- alpha를 0에 가깝게 만들어 훈련 세트의 성능은 나빠지지만 일반화에 도움을 주는 방법은 다음과 같습니다.

```python
ridge_10 = Ridge(alpha=10).fit(X_train, y_train)
print(f'훈련 세트 점수: {ridge_10.score(X_train, y_train):.2f}')
print(f'테스트 세트 점수: {ridge_10.score(X_test, y_test):.2f}')
```

- alpha를 조절하면 계수에 대한 제약이 바뀌며 아주 작은 alpha의 경우 LinearRegression과 거의 같아집니다.

```python
ridge_01 = Ridge(alpha=0.1).fit(X_train, y_train)
print(f'훈련 세트 점수: {ridge_01.score(X_train, y_train):.2f}')
print(f'테스트 세트 점수: {ridge_01.score(X_test, y_test):.2f}')
```

> 이 코드에서 alpha=0.1은 꽤 좋은 성능을 발휘합니다.

- 테스트 세트에 대한 성능이 높아질 때 까지 alpha 값을 줄일 수 있으며 이는 모델 복잡도와 관련됩니다.

- 또한 alpha 값에 따라 모델의 coef_ 속성이 어떻게 변화하는 지 조사한다면 alpha가 모델을 어떻게 변경하는지 더 깊게 이해할 수 있습니다.

    - 높은 alpha 값은 제약이 더 많으므로 낮을 때 보다 coef_의 절댓값 크기가 더 작을 것이라 예상할 수 있는데 이는 다음으로 확인할 수 있습니다.

```python
plt.plot(ridge_10.coef_, '^', label='Ridge alpha=10')
plt.plot(ridge.coef_, 's', label='Ridge alpha=1')
plt.plot(ridge_01.coef_, 'v', label='Ridge alpha=0.1')

plt.plot(lr.coef_, 'o', label='LinearRegression')
plt.xlabel('Coefficient List')
plt.ylabel('Coefficient Size')

xlims = plt.xlim()

plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-25, 25)
plt.legend()
```

> 위 그림에서 x(n번째 특성에 연관된 계수)축은 coef_ 원소 위치를, y축은 각 계수의 수치를 나타냅니다.

- alpha=10인 경우 대부분의 계수는 -3에서 3 사이에 위치하나 alpha=1일 때 더 커지고 0.1일 때 더욱 커지며 선형 회귀는 그림 밖으로 넘어가는 경우도 발생합니다.

- 규제 효과를 이해하는 또 다른 방법은 alpha를 고정하고 훈련 데이터의 크기를 변화시키는 방식입니다.

    - 다음은 보스턴 주택가격 데이터셋을 여러 크기로 샘플링해 학습한 것으로 데이터셋의 크기에 따른 모델의 성능 변화를 **학습 곡선**이라고 합니다.

```python
mglearn.plots.plot_ridge_n_samples()
```

- 예상대로 모든 데이터셋에 대해 리지와 선형 회귀 모두 훈련 세트의 점수가 테스트 세트의 점수보다 높습니다.

    - 리지는 규제가 적용되므로 리지의 훈련 데이터 점수가 전체적으로 선형 회귀보다 낮습니다.
    
    - 하지만 테스트 데이터는 리지의 점수가 높으며 데이터셋이 작을수록 그 차이는 더욱 커집니다.
    
        - 데이터가 400개 미만일 때 선형 회귀는 전혀 학습하지 못하며 두 모델 모두 데이터가 증가하면 성능도 증가합니다.(마지막에는 선형 회귀가 리지를 따라잡습니다.)

- 여기서 알 수 있는 것은 데이터가 충분해지면 규제가 덜 중요해져서 리지 회귀와 선형 회귀의 성능이 같아질 것이라는 점입니다.

- 또 다른 점은 선형 회귀의 훈련 데이터 성능이 감소하는 점으로 이는 데이터가 많아질수록 모델이 데이터를 기억하거나 과대적합하기 어려워지기 때문입니다.

#### 라소

- 선형 회귀에 규제를 적용하는 Ridge의 대안으로 **Lasso**가 있습니다.

    - 리지와 같이 라소도 계수를 0에 가깝게 만들지만 여기서는 L1 규제(계수의 절댓값의 합)를 사용합니다.
    
    - L1 규제를 사용하면 어떤 계수는 정말 0이 되며 모델에서 완전히 제외되는 특성이 발생합니다. 즉, 특성 선택이 자동으로 이루어집니다.

> 확장된 보스텀 주택가격 데이터셋에 라소를 적용하면 다음과 같습니다.

```python
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)
print(f'훈련 세트 점수: {lasso.score(X_train, y_train):.2f}')
print(f'훈련 세트 점수: {lasso.score(X_test, y_test):.2f}')
print('사용한 특성의 개수:', np.sum(lasso.coef_ != 0))
```

- 결과에서 볼 수 있듯이 Lasso는 훈련, 테스트 세트 모두 결과가 좋지 않습니다.

    - 이는 과소적합된 상태로 104개의 특성 중 4개만 사용된 것에서 알 수 있습니다.

- 이를 해결하기 위해 Ridge와 마찬가지로 Lasso도 계수를 얼마나 강하게 0으로 보낼 지 결정하는 `alpha` 매개변수를 지원합니다.

    - 과소적합을 줄이기 위해 alpha를 줄이고 대신 반복 실행하는 최대 횟수인 max_iter 변수를 키워야 합니다.

```python
# max_iter를 늘리지 않으면 이 값을 늘리라는 경고 메세지가 나옵니다.
lasso001 = Lasso(alpha=0.01, max_iter=100000).fit(X_train, y_train)
print(f'훈련 세트 점수: {lasso001.score(X_train, y_train):.2f}')
print(f'훈련 세트 점수: {lasso001.score(X_test, y_test):.2f}')
print('사용한 특성의 개수:', np.sum(lasso001.coef_ != 0))
```

- alpha 값을 낮추면 모델의 복잡도가 증가해 훈련 세트와 테스트 세트의 성능이 좋아집니다.

    - 성능이 Ridge 보다 조금 나아졌지만 사용된 특성이 104개 중 33개로 모델을 분석하기 쉽습니다.

- 하지만 alpha를 너무 낮추면 규제 효과가 없어져 과대적합이 되어 LinearRegression과 결과가 비슷해집니다.

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print(f'훈련 세트 점수: {lasso00001.score(X_train, y_train):.2f}')
print(f'훈련 세트 점수: {lasso00001.score(X_test, y_test):.2f}')
print('사용한 특성의 개수:', np.sum(lasso00001.coef_ != 0))
```

- 이제 이들을 비교하기 위해 alpha 값이 서로 다른 모델들의 계수를 그래프로 그려보면 다음과 같습니다.

```python
plt.plot(lasso.coef_, 's', label='Lasso alpha=1')
plt.plot(lasso001.coef_, '^', label='Lasso alpha=0.01')
plt.plot(lasso00001.coef_, 'v', label='Lasso alpha=0.0001')

plt.plot(ridge_01.coef_, 'o', label='Ridge alpha=0.1')
plt.legend(ncol=2, loc=(0, 1.05))
plt.ylim(-25, 25)
plt.xlabel('Coefficient List')
plt.ylabel('Coefficient Size')
```

- alpha=1일 때 계수 대부분이 0이며 나머지 계수들도 크기가 매우 작습니다.

- alpha=0.01일 때 대부분의 특성이 0이 되며 0.0001로 줄이면 값이 커져 꽤 규제받지 않은 모델이 나옵니다.

    - 이와 비슷하게 리지는 alpha=0.1인 경우 alpha=0.01인 라소와 비슷하지만 Ridge는 어떤 계수도 0이 되지 않습니다.

- 실제 두 모델 중 보통 리지 회귀가 선호됩니다. 하지만 특성이 많으면서 그 중 일부분만 중요하다면 Lasso가 더 나을 수 있습니다.

- 또한 분석하기 쉬운 모델을 원한다면 Lasso가 입력 특성 중 일부만 사용하므로 더 쉽게 해석할 수 있습니다.

- 한편, scikit-learn은 Lasso와 Ridge를 결합한 ElasticNet도 제공합니다. 이 모델을 최상의 성능을 내지만 L1 규제와 L2 규제를 위한 매개변수를 모두 조정해야 합니다.

#### 분류용 선형 모델

- 선형 모델은 분류에도 널리 사용되는데 그 중 하나인 이진 분류를 살펴보겠습니다.

    - 분류용 선형 모델은 선형 회귀와 식이 유사하지만 특성들의 가중치의 합을 임계치 0과 비교합니다.
    
    - 즉, 0보다 작은 값은 클래스 -1, 큰 값은 클래스 1로 예측합니다. 이 규칙은 분류에 쓰이는 모든 선형 모델이 동일합니다.
    
    - 이 방식도 계수와 절편을 찾기 위한 다양한 방법이 있습니다.

- 회귀용 선형 모델에서 출력 y의 특성이 선형 함수, 즉 직선, 평면, 초평면(차원이 3 이상인 경우)인 반면에 분류용 선형 모델은 **결정 경계**가 입력의 선형 함수입니다. 다시 말해 (이진) 선형 문류기는 선, 평면, 초평면을 이용해 두 개의 클래스를 구분하는 분류기입니다.

- 선형 모델을 학습시키는 알고리즘은 다양하나 크게 다음 두 가지로 분류할 수 있습니다.

    1. 특정 계수와 절편의 조합이 훈련 데이터에 얼마나 잘 맞는지 측정하는 방법
    
    2. 사용할 수 있는 규제가 있는지와 있다면 어떤 방식으로 규제하는지

- 알고리즘들은 훈련 세트를 잘 학습하는지 측정하는 방법이 각기 다르지만 불행히도 수학적, 기술적인 이유로 알고리즘들이 만드는 잘못된 분류를 최소화하기 위해 w와 b를 조정하는 것은 불가능합니다.

    - 즉, 많은 애플리케이션에서 앞 목록의 첫 번째 항목인 **손실 함수**의 차이는 크게 중요하지 않습니다.

- 가장 널리 알려진 두 개의 선형 분류 알고리즘은 **로지스틱 회귀** 방식과 **서포트 벡터 머신** 방식이 존재합니다. 여기서 로지스틱 회귀는 이름에 회귀가 들어가나 분류 알고리즘임에 유념해야 합니다.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

X, y = mglearn.datasets.make_forge()

fig, axes = plt.subplots(1, 2, figsize=(10, 3))
for model, ax in zip([LinearSVC(), LogisticRegression()], axes):
    clf = model.fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=False, eps=0.5, ax=ax, alpha=0.7)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title(clf.__class__.__name__)
    ax.set_xlabel('attribute 0')
    ax.set_ylabel('attribute 1')
axes[0].legend()
```

- 이 그림은 이전처럼 forge 데이터셋의 첫 번째 특성을 x축에, 두 번째 특성을 y축에 놓고 각 모델로 만든 결정 경계가 직선으로 표현되었으며 위 아래로 클래스 1과 클래스 0을 구분하고 있습니다.

    - 달리 말해 새로운 데이터가 직선 위에 놓인다면 클래스 1, 아니면 클래스 0이 될 것입니다.

- 두 모델은 비슷한 결정 경계를 만들었고 똑같이 포인트 두 개를 잘못 분류했습니다. 이는 회귀에서 본 Ridge와 바찬가지로 두 모델이 L2 규제를 기본적으로 사용하기 때문입니다.

- LogisticRegression과 LinearSVC에서 규제의 강도를 결정하는 매개변수는 C로 C가 높아지면 규제가 감소합니다.

    - 다시 말해 높은 C는 각 모델이 훈련 세트에 최대한 가깝게 맞추려 하고 낮은 C는 계수 백터가 0에 가까워지도록 만듭니다.

    - 매개변수 C의 작동 방식을 다르게 설명하자면, C를 낮추면 데이터 포인트 중 다수에 맞추려고 하지만 C를 높이면 데이터 포인트를 정확히 분류하려고 할 것입니다.

```python
# LinearSVC의 C값에 따른 동작은 다음과 같습니다.
mglearn.plots.plot_linear_svc_regularization()
```

1. 왼쪽은 아주 작은 C로 인해 규제가 많이 적용되어 클래스 0의 대부분이 아래, 1이 위에 존재합니다.

    - 규제가 강한 모델은 비교적 수평에 가까운 결정 경계를 만들고 두 개의 데이터 포인트를 잘못 분류했습니다.

2. 중간은 C가 조금 더 커져 잘못 분류한 두 샘플에 민감해져 결정 경계가 기울어져 있습니다.

3. 오른쪽은 C를 매우 키워 결정 경계가 더 기울었고 클래스 0의 데이터 포인트를 모두 완벽히 분류했습니다.

    - 다만 직선으로 분류하는 것은 한계가 있기 때문에 클래스 1의 포인트 하나가 잘못 분류되어 있습니다.
    
    - 즉, 오른쪽은 모든 데이터 포인트를 정확히 분류하려 했으나 클래스의 전체적인 배치를 잘 파악하지 못한 과대적합된 상태입니다.

- 회귀와 비슷하게 분류 선형 모델은 낮은 차원의 데이터에서는 결정 경계가 직선이거나 평면이기에 매우 제한적인 것처럼 보입니다.

- 하지만 고차원에서는 분류 선형 모델이 매우 강력해지며 특성이 많아질 경우 과대적합되지 않도록 하는 것이 중요합니다.

> 유방암 데이터셋을 통해 LogisticRegression을 더 자세히 분석하겠습니다.

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42)

logreg = LogisticRegression(max_iter=100000).fit(X_train, y_train)
print(f'훈련 세트 점수: {logreg.score(X_train, y_train):.3f}')
print(f'테스트 세트 점수: {logreg.score(X_test, y_test):.3f}')
```

- 기본값이 C=1이기에 훈련과 테스트 모두 95% 정확도의 훌륭한 성능을 내고 있습니다.

- 하지만 훈련 세트와 테스트 세트의 성능이 매우 비슷하므로 과소적합 상태인 듯하여 모델의 제약을 풀어주기 위해 C를 증가시키겠습니다.

```python
logreg100 = LogisticRegression(C=100, max_iter=100000).fit(X_train, y_train)
print(f'훈련 세트 점수: {logreg100.score(X_train, y_train):.3f}')
print(f'테스트 세트 점수: {logreg100.score(X_test, y_test):.3f}')
```

- C를 증가시키니 훈련 세트의 정확도가 높아지고 테스트 세트도 약간 상승했습니다. 즉, 복잡한 모델일수록 성능이 좋아짐을 의미합니다.

- 반대로 규제를 강하게 C=0.01을 사용한 결과는 아래와 같습니다.

```python
logreg001 = LogisticRegression(C=0.01, max_iter=100000).fit(X_train, y_train)
print(f'훈련 세트 점수: {logreg001.score(X_train, y_train):.3f}')
print(f'테스트 세트 점수: {logreg001.score(X_test, y_test):.3f}')
```

- 과소적합된 상태에서 더 왼쪽으로 이동하므로 훈련 세트와 테스트 세트의 정확도가 기본 매개변수보다 낮아집니다.

- 그럼 이 매개변수 설정을 다르게 학습시킨 모델의 계수를 확인해보겠습니다.

```python
plt.plot(logreg100.coef_.T, '^', label='C=100')
plt.plot(logreg.coef_.T, 'o', label='C=1')
plt.plot(logreg001.coef_.T, 'v', label='C=0.01')
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-5, 5)
plt.xlabel('attributes')
plt.ylabel('coefficient size')
plt.legend()
```

#### Note

- LogisticRegression은 기본으로 L2 규제를 적용하므로 Ridge의 모습과 유사합니다.

    - 즉, 규제를 강하게 할수록 계수들은 0에 가까워지나 완전히 0이 되지 않습니다.

- 위 그림에서 재밌는 현상을 볼 수 있는데 C=100, C=1일 때 세 번째(mean preimeter) 계수는 음수지만 반대에선 양수로 C=1일 때 보다도 절댓값이 커집니다.

- 이를 해석하면 계수와 클래스의 상관관계를 알려줄 수 있습니다.

    - 예를 들어 texture error은 악성 샘플과 관련이 깊지만 mean perimeter는 양성이나 악성 모두의 신호가 될 수 있습니다.

- 그래서 선형 모델의 계수는 항상 의심해야 합니다.

- 더 이해하기 쉬운 모델을 위해서는 L1 규제를 사용하는 것이 좋습니다. 다음은 L1 규제일 때 분류 정확도와 계수입니다.

```python
# 무언가... 무언가 틀렸다...
for C, marker in zip([0.001, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(C=C, penalty='l1', max_iter=100000, solver='saga').fit(X_train, y_train)
    print(f'훈련 세트 점수: {lr_l1.score(X_train, y_train):.3f}')
    print(f'테스트 세트 점수: {lr_l1.score(X_test, y_test):.3f}')
    plt.plot(lr_l1.coef_.T, marker, label=f'C={C:.3f}')

plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-5, 5)
plt.xlabel('attributes')
plt.ylabel('coefficient size')
plt.legend(loc=3)
```

- 이처럼 이진 분류 선형 모델과 회귀 선형 모델 사이에는 유사점이 많으며 회귀처럼 모델들의 주요 차이인 규제에서 모든 특성을 사용할 것인지 아닌지 결정하는 변수가 penalty 매개변수입니다.

#### 다중 클래스 분류용 선형 모델

- 로지스틱 회귀를 제외하고 많은 선형 분류 모델은 태생적으로 이진 분류만을 지원합니다. 즉, 다중 클래스를 지원하지 않습니다.

- 이진 분류 알고리즘을 다중 클래스 분류 알고리즘으로 확장하는 보편적인 기번은 일대다 기법으로 이를 기반으로 각 클래스를 다른 모든 클래스와 구분하도록 이진 분류 모델을 학습시킵니다.

    - 결국 클래스 수만큼 이진 분류 모델이 만들어지며 예측 시 모든 이진 분류기가 작동해 가장 높은 점수를 내는 분류기의 클래스 값으로 선택됩니다.

    - 즉, 클래스별로 이진 분류기를 만들면 각 클래스가 계수 벡터와 절편을 가지게 되며 분류 신뢰도를 나타내는 값이 가장 높은 클래스가 해당 데이터의 클래스가 됩니다.

- 다중 클래스 로지스틱 회귀 이면의 수학은 일대다 방식과 조금 다르지만 하나의 클래스마다 하나의 계수 벡터와 절편을 만들며 예측 방법도 같습니다.

> 세 개의 클래스를 가진 간단한 데이터셋에 일대다 방식을 적용하면 다음과 같습니다.
>
> 이 데이터셋은 2차원으로 각 클래스의 데이터가 정규분포를 따릅니다.

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel('attribute 0')
plt.xlabel('attribute 1')
plt.legend(['class 0', 'class 1', 'class 2'])
```

```python
# 위 데이터셋으로 LinearSVC 분류기를 훈련해보겠습니다.
linear_svm = LinearSVC().fit(X, y)
print('계수 배열의 크기:', linear_svm.coef_.shape)
print('절편 배열의 크기:', linear_svm.intercept_.shape)
```

- coef_ 배열은 (3, 2)로 세 개의 클래스에 각각 대응하는 계수 벡터를 담고 있으며 열은 각 특성에 따른 계수 값을 가지고 있습니다.

- 반면, intercept_는 각 클래스의 절편을 담은 1차원 벡터입니다.

> 이제 세 이진 분류기가 만드는 경계를 시각화하겠습니다

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 15)
plt.xlabel('attribute 0')
plt.xlabel('attribute 1')
plt.legend(['class 0', 'class 1', 'class 2', 'class 0 barrier', 'class 1 barrier', 'class 2 barrier'], loc=(1.01, 0.3))
```

- 훈련 데이터의 클래스 0에 속한 모든 포인트를 클래스 0을 구분하는 직선 내부에 존재하며 다른 두 클래스와 구분하는 직선 외부에 존재해 다른 클래스와 구별됨을 보여줍니다.

    - 이는 클래스 1, 클래스 2도 마찬가지로 다른 클래스와 다름이 분명히 구별됩니다.

- 하지만 중앙의 삼각형의 경우 세 분류기가 모두 나머지 영역으로 분류했는데 이 경우 분류기의 직선과 가장 가까운 클래스의 클래스로 구별됩니다.

> 위 결과의 각 영역에 색을 칠하면 더 분명히 알 수 있을 것입니다.

```python
mglearn.plots.plot_2d_classification(linear_svm, X, fill=True, alpha=0.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, mglearn.cm3.colors):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.xlabel('attribute 0')
plt.xlabel('attribute 1')
plt.legend(['class 0', 'class 1', 'class 2', 'class 0 barrier', 'class 1 barrier', 'class 2 barrier'], loc=(1.01, 0.3))
```

#### 장단점과 매개변수

- 선형 모델의 주요 매개변수는 회귀 모델에서는 alpha, LinearSVC와 LogisticRegression에서는 C입니다.

    - alpha가 클수록, C가 작을수록 모델은 단순해지고 특히 회귀 모델에서 이 매개변수를 조절하는 것은 매우 중요합니다.

- 보통 alpha와 C는 로그 스케일로 최적치를 정하며 L1 규제를 사용할 지, L2 규제를 사용할 지도 정해야 합니다.

    - 중요 특성에 맞지 않는다고 생각되면 L1을, 그렇지 않으면 기본적으로 L2 규제를 사용하는 게 좋습니다.
    
- L1 규제는 모델의 해석의 중요 요소로 몇 가지 특성만 사용하기 때문에 해당 모델에 중요한 특성이 무엇인지 그 효과가 어느 정도인지 알 수 있습니다.

- 선형 모델은 학습 속도와 예측이 빠르며 매우 큰 데이터셋과 희소한 데이터셋도 가리지 않습니다.

    - 수십만 이상의 대용량 데이터셋이라면 기본 설정보다 빨리 처리하도록 LogisticRegression과 Ridge에서 `solver='sag'` 옵션을 주면 됩니다.
    
    - 다른 대안으로는 선형 모델의 대용량 처리 버전인 SGDClassifier와 SGDRegressor를 사용해도 됩니다.

- 선형 모델의 또 하나의 장점은 회귀와 분류에서 본 공식을 사용해 예측이 어떻게 만들어질 지 쉽게 이해할 수 있다는 것입니다.

- 하지만 계수의 값이 왜 그렇게 나오는지 명확하지 않을 때가 종종 있음 특히 데이터셋의 특성들이 서로 깊게 연관되어 있을 때 더욱 그렇습니다. 그리고 이 경우 계수를 분석하기 매우 어려울 수 있습니다.

- 선형 모델은 샘플에 비해 특성이 많을 때 잘 작동하여 다른 모델로 학습하기 어려운 매우 큰 데이터셋에도 선형 모델을 많이 사용합니다.

    - 그러나 저차원의 데이터셋은 다른 모델들의 일반화 성능이 더 좋습니다.

#### 메서드 연결

- 모든 scikit-learn의 fit 메서드는 self를 반환하므로 다음과 같은 코드를 작성할 수 있습니다.

```python
from sklearn.linear_model import LogisticRegression

# one line command로 모델 객체의 생성과 학습을 동시에 할 수 있습니다.
logreg = LogisticRegression().fit(X_train, y_train)
```

- fit 메서드의 반환값은 학습된 모델로 정의한 변수에 할당됩니다. 이처럼 메서드 호출을 잇는 것을 **메서드 연결**이라고 합니다.

- scikit-learn에서 fit과 predict도 한 줄에 쓰느 메서드 연결이 가능합니다.

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train).predict(X_test)
```

- 궁극적으로는 모델 객체 생성, 훈련, 예측을 모두 one line command로 실행할 수 있습니다.

> 이렇게 짧게 쓰는 것은 한 줄에 메서드가 많이 들어가 코드를 읽기 힘들어져 바람직하진 않습니다.
>
> 더군다나 학습된 모델 자체를 변수에 할당하지 않아 다른 데이터에 대해 예측하거나 모델을 분석할 수 없습니다.

```python
y_pred = LogisticRegression().fit(X_train, y_train).predict(X_test)
```

### 2.3.4 나이브 베이즈 분류기

- 나이브 베이즈 분류기는 앞 절의 선형 모델, 즉 LogisticRegression이나 LinearSVC 같은 선형 분류기와 유사하지만 훈련 속도가 더 빠르고 일반화 성능이 조금 뒤집니다.

- 나이브 분류기가 효과적인 이유는 각 특성을 개별로 취급해 파라미터를 학습하고 각 특성에서 클래스별 통계를 단순하게 취합하기 때문입니다.

    - scikit-learn에 구현된 나이브 베이즈 분류기는 GaussionNB, BernoulliNB, MultinomialNB가 있으며 순서대로 연속적인 데이터, 이진 데이터, 카운트 데이터(특성이 어떤 것을 헤아린 정수로만 구성된 데이터, 예: 문장 내 단어의 횟수)에 적용됩니다.
    
    - BernoulliNB, MultinomialNB는 대부분 텍스트 데이터를 분류할 때 사용합니다.

- BernoulliNB 분류기는 각 틀래스의 특성 중 0이 아닌 것이 몇 개인지 세는 방식으로 다음과 같습니다.

```python
X = np.array([[0, 1, 0, 1],
              [1, 0, 1, 0],
              [0, 0, 0, 1],
              [1, 0, 1, 1]])
y = np.array([0, 1, 0, 1])
```

- 이진 특성을 4개 가진 데이터 포인트 4개가 있고 클래스는 0과 1 두 개일 때 출력 y의 클래스가 0인 경우 첫 번째 특성과 세 번째 특성이 모두 0인 경우는 X의 0번, 3번 데이터 포인트입니다.

- 이처럼 y의 클래스가 1인 경우도 계산하면 클래스별로 원소를 셀 수 있고 이를 다음과 같이 표현합니다.

```python
cnt = {}
for label in np.unique(y):
    # 각 클래스마다 반복을 해서 1이 나타난 갯수를 각 열에 대해 셉니다.
    cnt[label] = X[y == label].sum(axis=0)
print('특성 카운트:', cnt)
```

- GaussionNB와 MultinomialNB는 계산하는 통계 데이터의 종류가 조금 다릅니다.

    - MultinomialNB는 클래스별로 특성의 평균을, GaussionNB는 클래스별로 각 특성의 표준편차와 평균을 저장합니다.

- 예측할 땐 데이터 포인트를 클래스의 통계 값과 비교해 가장 잘 맞는 클래스를 예측값으로 합니다.

    - BernoulliNB와 MultinomialNB의 예측 공식은 선형 모델과 형태가 같습니다.
    
    - 그러나 나이브 베이즈 모델의 coef_는 기울기 w가 아니라 선형 모델가 조금 다릅니다.

#### 장단점과 매개변수

- BernoulliNB와 MultinomialNB은 모델의 복잡도는 조절하는 alpha 매개변수를 가지고 있습니다.

    - alpha에 따라 모든 특성에 양의 값을 가진 가상의 데이터 포인트를 alpha 갯수 만큼 추가하는 알고리즘이 적용됩니다.
    
    - 이는 곧 통계 데이터를 완만하게 만들어지고 그 정도는 alpha와 비례합니다.

- alpha에 따른 알고리즘 성능 변동은 크지 않아 성능 향상에 크게 기여하지 않지만 이 값을 조정해 어느 정도 정확한 값을 얻을 수 있습니다.

- GaussianNB는 대부분 매우 고차원인 데이터셋에 사용하고 다른 모델들은 텍스트 같은 희소 데이터를 카운트하는 데 사용됩니다.

    - MultinomialNB는 보통 0이 아닌 큭성이 비교적 많은 데이터셋에서 BernoulliNB보다 성능이 높습니다.

- 나이브 베이즈 모델과 선형 모델의 장단점은 비슷합니다.

    - 훈련과 예측 속도가 빠르며 훈련 과정을 이해하기 쉽습니다. 희소한 고차원 데이터에서 잘 작동하며 비교적 매개변수에 민감하지 않습니다.
    
    - 선형 모델로 오래 걸리는 매우 큰 데이터셋은 나이브 베이즈 모델을 시도해볼 만합니다.

#### 결정 트리 만들기

- 2차원 데이터셋을 분류하는 결정 트리를 만들기 위해 각 클래스에 50개의 데이터 포인트가 있고 반달 두 개가 쪼개진 듯한 모양을 가진 two_moons 데이터셋을 사용해보겠습니다.

- 결정트리를 학습한다는 것은 정답에 가장 빨리 도달하는 예/아니오 질문 목록을 학습한다는 뜻으로 머신러닝에서 이 질문들을 **테스트**라고 합니다.(모델이 잘 일반화되었는지 확인하는 테스트 세트가 아닙니다!)

    - 보통 데이터는 앞 선 예시처럼 예/아니오 형태가 아닌 2차원 데이터셋처럼 연속된 특성으로 구성됩니다. 즉, '특성 i는 값 a보다 큰가?'와 같은 형태입니다.

- 트리를 만들 때 알고리즘은 다능한 모든 테스트에서 타깃값에 대해 가장 많은 정보를 가진 것을 고릅니다.

    - 처음에 가장 많은 정보를 포함하는 경계를 설정해 클래스를 구분시킵니다. 이 때 **루트 노드**라고 불리는 맨 위 노드에서 전체 데이터셋을 구분합니다.
    
    - 이제 이 기준을 기반으로 각 클래스에 몇 개의 데이터셋이 포함되어있는지 확인합니다. 이 때 분류가 완벽하지 않다면 다음 직선을 기준으로 데이터셋을 구분하고 이를 또 검사합니다.
    
- 이 프로세스를 반복하면 각 노드가 테스트 하나씩을 가진 이진 결정 트리를 만듭니다. 즉, 각 테스트는 하나의 축을 따라 데이터를 둘로 나눕니다.

    - 이는 계층적으로 영역을 분할하는 알고리즘으로 여길 수 있으며 각 테스트는 하나의 특성에 대해서만 이루어져 나뉜 영역은 항상 축에 평행합니다.

- 데이터를 분할하는 것은 각 분할된 영역이 한 개의 타깃값(즉, 클래스나 회귀 결과)를 가질 때까지 반복하며 타깃 하나로만 이루어진 노드를 **순수 노드**라고 합니다.

- 새로운 데이터 포인트에 대한 예측은 주어진 데이터 포인트가 분할된 특성 영역 중 어디에 놓이는 지 확인하면 됩니다.

    - 그래서 그 영역의 타깃값 중 다수인 것을 예측 결과로 사용합니다. 이는 루트 노드부터 왼쪽 혹은 오른쪽으로 탐색하며 확인할 수 있습니다.

- 같은 방법으로 회귀 문제에도 트리를 사용할 수 있습니다.

    - 예측을 하려면 각 노드의 테스트 결과에 따라 트리를 탐색하고 새로운 데이터 포인트에 해당되는 리프 노드를 찾습니다.
    
    - 찾은 리프 노드의 훈련 데이터 평균값이 이 데이터 포인트의 출력이 됩니다.

#### 결정 트리의 복잡도 제어하기

- 일반적으로 트리 만들기를 모든 리프 노드가 순수 노드가 될 때까지 진행하면 모델이 매우 복잡해지고 훈련 데이터에 과대적합됩니다.

    - 즉, 훈련 세트의 모든 데이터 포인트가 정확한 클래스의 리프 노드가 되는 상태로 이는 바람직한 결정 경계의 모습이 아닙니다.

- 과대적합을 막는 전략은 크게 두 가지로 트리 생성을 일찍 중단시키는 **사전 가지치기**와 데이터 포인트가 적은 노드를 삭제하거나 병합하는 **사후 가지치기** 혹은 **가지치기**입니다.

    - 사전 가지치기는 트리의 최대 깊이, 리프의 최대 개수, 분할 포인트의 최소 갯수 등을 지정하는 방법으로 실행합니다.

    - scikit-learn은 DecisionTreeRegressor와 DecisionTreeClassifier에 구현되어 있으며 사전 가지치기만 지원합니다.

- 유방암 데이터셋을 활용해 사전 가지치기의 효과를 알아보겠습니다. 항상 해왔던 방법들을 진행하고 모델은 완전 트리 모델과 random_state 옵션을 준 모델을 비교합니다.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=42)
tree = DecisionTreeClassifier(random_state=0)
tree.fit(X_train, y_train)
print(f'훈련 세트 정확도: {tree.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {tree.score(X_test, y_test):.3f}')
```

- 모든 리프 노드가 순수 노드가 되어 훈련 세트의 정확도가 100%가 되고 테스트 세트의 정확도응 95%보다 낮아졌습니다.

- 이처럼 결정 트리의 깊이를 제한하지 않으면 트리는 무한정 깊어지고 복잡해질 수 있어 과대적합되기 쉽고 새로운 데이터에 잘 일반화되지 않습니다.

- 이제 사전 가지치기를 통해 트리의 성장을 막을 것입니다. max_depth 옵션을 통해 이를 지정할 수 있으며 이를 통해 연속 질문의 갯수를 제한합니다.

    - 이는 훈련 세트의 정확도는 떨어트리지만 테스트 세트의 정확도는 개선합니다.

```python
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)
print(f'훈련 세트 정확도: {tree.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {tree.score(X_test, y_test):.3f}')
```

#### 트리의 특성 중요도

- 전체 트리를 살펴보는 것이 어려울 수 있으니 트리가 어떻게 작동하는지 요약하는 속성을 사용할 수 있습니다.

    - 가장 널리 사용되는 속성은 트리를 만드는 결정에 각 특성이 얼마나 중요한지를 평가하는 **특성 중요도**입니다.
    
    - 이 값은 0과 1사이의 숫자로 높을수록 중요함을 의미합니다. 특성 중요도의 전체 합은 1이 됩니다.

```python
print('특성 중요도:\n', tree.feature_importances_)
```

- 선형 모델의 계수를 시각화하는 것도 비슷한 방법으로 특성 중요도를 시각화할 수 있습니다.

```python
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(np.arange(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel('Attribute Importance')
    plt.ylabel('Attribute')
    plt.ylim(-1, n_features)

plot_feature_importances_cancer(tree)
```

- 첫 번째 노드에서 사용한 특성인 worst radius가 가장 중요한 특성으로 나타나며 이는 첫 노드에서 두 클래스를 잘 나누고 있음을 방증합니다.

- 그러나 어떤 특성의 feature_importances_ 값이 낮다고 이 특성이 유용하지 않음을 말하지는 않습니다.

    - 단지 트리가 해당 특성을 선택하지 않았고 다른 특성이 비슷한 정보를 가지고 있음을 의미할 수도 있습니다.

- 선형 모델의 계수와 달리 특성 중요도는 항상 양수로 특성이 어떤 클래스를 지지하는지 알 수 없습니다.

    - 즉, worst radius가 중요하지만 이것이 양성인지 악성인지를 의미하지는 않습니다. 이는 특성과 클래스 사이에 간단하지 않은 관계가 있을 수 있다는 점을 의미하며 다음 예시로 이를 살펴보겠습니다.

    - 회귀 트리는 분류 트리와 사용과 분석이 비슷하나 회귀를 위해 트리 기반 모델을 사용할 때 **외삽**, 즉 훈련 데이터 범위 외의 포인트에 대해 예측할 수 없음을 알아야합니다.

> 컴퓨터 메모리 가격 동향 데이터셋을 이용해 더 자세히 이를 설명해보겠습니다.

```python
import os

ram_prices = pd.read_csv(os.path.join(mglearn.datasets.DATA_PATH, 'ram_price.csv'))

plt.yticks(fontname='Arial')
plt.semilogy(ram_prices.date, ram_prices.price)
plt.xlabel('year')
plt.ylabel('price ($/MB)')
```

- 위 그래프는 y축이 로그 스케일로 약간의 굴곡을 제외하고 선형적으로 나타납니다.

- 날짜 특성만으로 2000년 전까지 데이터를 바타응로 2000년 후의 가격을 예측해보겠습니다.

    - 여기서는 간단한 두 모델 DecisionTreeRegressor와 LinearRegression을 사용할 것입니다.
    
    - 가격이 로그 스케일이므로 비교적 선형 관계를 가질 것이며 트리 모델은 사용에 문제가 없지만 선형 모델을 차이가 있습니다.
    
        - 이를 위해 훈련과 예측 후 로그 스케일로 돌리는 지수 함수를 적용할 것입니다.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression

# 2000년 이전 데이터를 훈련 데이터로, 그 외를 테스트 데이터로 사용하겠습니다.
data_train = ram_prices[ram_prices.date < 2000]
data_test = ram_prices[ram_prices.date >= 2000]

# 가격 예측은 날짜만을 사용합니다.
X_train = data_train.date[:, np.newaxis]
# 데이터와 타깃 사이의 관계를 간단히 하기 위해 로그 스케일로 바꾸겠습니다.
y_train = np.log(data_train.price)

tree = DecisionTreeRegressor().fit(X_train, y_train)
linear_reg = LinearRegression().fit(X_train, y_train)

# 예측은 전체 기간에 대해 수행합니다.
X_all = ram_prices.date[:, np.newaxis]

pred_tree = tree.predict(X_all)
pred_lr = linear_reg.predict(X_all)

# 예측 값의 로그 스케일을 복구합니다.
price_tree = np.exp(pred_tree)
price_lr = np.exp(pred_lr)
```

```python
# 이제 결과를 비교하겠습니다.
plt.semilogy(data_train.date, data_train.price, label='train data')
plt.semilogy(data_test.date, data_test.price, label='test data')
plt.semilogy(ram_prices.date, price_tree, label='tree predict')
plt.semilogy(ram_prices.date, price_lr, label='linear reg predict')
plt.legend()
```

- 두 모델을 확연한 차이를 보입니다. 선형 모델은 우리가 아는대로 직선으로 데이터를 근사합니다.

    - 이 직선은 데이터에 있는 미세한 굴곡을 매끈하게 근사하여 테스트 데이터를 꽤 정확히 예측합니다.

- 반면에 트리 모델은 훈련 데이터를 완벽하게 예측합니다. 하지만 모델이 가진 데이터 범위 밖의 데이터는 예측에 실패합니다.

    - 훈련 데이터는 트리 복잡도에 제한을 두지 않아 모든 데이터셋을 기억하기 때문입니다. 하지만 테스트 데이터는 단순히 마지막 포인트를 사용하기 때문에 이런 현상이 발생합니다.

- 이는 모든 트리 기반 모델의 공통된 단점입니다.

#### 장단점과 매개변수

- 결정 트리에서 모델 복잡도를 조절하는 매개변수는 사전 가지치기 변수로 max_depth, max_leaf_nodes, min_samples_leaf 중 하나만 지정해도 과대적합을 막기에 적합합니다.

- 결정 트리가 다른 알고리즘보다 나은 점은 두 가지입니다.

    1. 만들어진 모델을 쉽게 시각화해 이해하기 쉽습니다.
    
    2. 데이터의 스케일에 구애받지 않습니다.
    
        - 각 특성이 개별적으로 처리되어 데이터를 분할하는데 데이터 스케일의 영향을 받지 않으므로 결정 트리는 정규화나 표준화같은 전처리 과정이 필요하지 않습니다.
        
        - 특히, 특성 스케일이 서로 다르거나 이진, 연속 특성이 혼합되어 있어도 잘 작동합니다.

- 반면 단점은 사전 가지치기를 사용해도 과대적합되는 경향이 있어 일반화 성능이 좋지 않다는 점입니다.

    - 이를 해결하기 위해서 앙상블 방법을 단일 결정 트리의 대안으로 흔히 사용합니다.

### 2.3.6 결정 트리의 앙상블

- **앙상블**이란 여러 머신러닝 모델을 연결하여 더 강력한 모델을 만드는 기법입니다.

    - 머신러닝은 이런 종류의 모델이 많지만 그 중 다음 두 앙상블 모델이 분류와 회귀 문제의 다양한 데이터셋에 효과적이라고 입증되었습니다. 바로 **랜덤 포레스트**와 **그레이디언트 부스팅**입니다.

#### 랜덤 포레스트

- 앞선 결정 트리의 주요 단점인 훈련 데이터에 과대적합되는 경향을 해결하는 방법 중 하나입니다.

- 랜덤 포레스트는 조금씩 다른 여러 결정 트리의 묶음으로 각 트리가 예측을 잘 하나 데이터에 과대적합되는 경향을 가진다는 점에서 출발합니다.

    - 이를 위해 서로 다른 방향으로 과대적합된 트리를 많이 만들고 이를 평균애면 과대적합된 양을 줄일 수 있는데 이를 통해 예측 성능을 유지하면서 과대적합을 줄일 수 있습니다.

- 이를 위해서 결정 트리가 많이 필요하며 각 트리는 예측 성능이 좋아야하며 다른 트리와는 구별되야 합니다.

    - 이처럼 랜덤 포레스트는 트리 생성 시 무작위성을 주입하는데 트리를 만들 때 사용하는 데이터 포인트를 무작위로 선택하는 방법과 분할 테스트에서 특성을 무작위로 선택하는 방법이 있습니다.

#### 랜덤 포레스트 구축

- 랜덤 포레스트 모델을 만드려면 생성할 트리의 개수를 정해야하는데 RandomForestRegressor나 RandomForestClassifier의 n_estimators 매개변수를 조절하여 결정합니다.

    - 트리들이 완전히 독립적으로 만들어져야 하므로 각 트리가 고유하게 만들어지는 무작위 선택을 하는데 여기서 **부트스트랩 샘플**을 생성합니다.
    
    - 다시 말해 n_samples개의 데이터 포인트 중 무작위로 데이터를 n_samples만큼 반복 추출하여 원본 데이터셋과 크기는 같지만 일부 중복 혹은 누락이 있는 데이터셋이됩니다.

- 이렇게 만든 데이터셋으로 결정 트리를 만드는데 각 노드에서 전체 특성을 대상으로 최선의 테스트를 찾는 결정 트리가 아닌 후보 특성을 무작위로 선택하고 이들 중 최선의 테스트를 찾습니다.

    - 몇 개의 특성을 고를지는 max_features로 조절하며 고르는 것은 매 노드마다 반복되므로 트리의 각 노드는 다른 후보 특성들을 사용해 테스트를 만듭니다.

- 부트스트랩 샘플링은 랜덤 포레스트 트리가 조금씩 다른 데이터셋을 이용해 만들어지도록하며 각 노드가 특성의 일부만 사용하므노 트리의 각 분기는 각각 다른 특성 부분 집합을 사용하게됩니다. 이를 거쳐 랜덤 포레스트의 모든 트리가 서로 달라지게 됩니다.

- 이 방식의 핵심 매개변수는 max_features로 해당 값이 n_features가 되면 모든 특성을 고려하므로 선택에 무작위성이 들어가지 않습니다.

    - 반면에 1로 하면 테스트할 특성을 고를 필요 없이 무작위로 선택한 특성의 임계값을 찾습니다.
    
    - 즉, max_features가 커질수록 랜덤 포레스트 트리들이 비슷해지고 데이터에 잘 맞춰지며 낮아질수록 트리들이 달라지고 트리의 깊이가 깊어집니다.

- 랜덤 포레스트로 예측을 할 때는 먼저 알고리즘이 모델에 있는 모든 트리의 예측을 만들고 회귀의 경우 이 예측들을 사용해 평균으로 최종 예측을 만듭니다.

    - 분류의 경우에는 약한 투표 전략, 즉 각 알고리즘이 출력 레이블의 확률을 제공하여 간접 예측을 하고 이를 평균내어 가장 높은 확률을 가진 클래스가 예측값이 됩니다.

#### 장단점과 매개변수

- 회귀와 분류에 있어 랜덤 포레스트는 가장 널리 사용되는 머신러닝 알고리즘으로 성능이 뛰어나고 매개변수 튜닝을 많이 하지 않아도 잘 작동하며 데이터의 스케일을 맞출 필요가 없습니다.

- 기본적으로 랜덤 포레스트는 단일 트리의 단점을 보완하고 장점을 그대로 가지고 있습니다.

    - 만일 의사 결정 과정을 간소하게 표현한다면 단일 트리를 사용할 수 있는데 수십 개 이상의 트리를 자세히 분석하는 것은 어렵고 단일 결정 트리보다 랜덤 포레스트 트리가 더 깊어지는 경향이 있기 때문입니다.
    
    - 대량의 데이터셋에서 랜덤 포레스트 분석 시간이 걸릴 수 있지만 멀티 코어 프로세서를 사용하면 n_jobs를 지정해 병렬 처리를 하여 더 빠르게 분석할 수 있습니다.(n_jobs=-1은 모든 코어를 사용함을 의미합니다.)

    - 트리가 많을수록 random_state의 변화에 따른 결과 변화도 적지만 같은 결과를 원한다면 이 값을 고정해야 합니다.

- 랜덤 포레스트는 텍스트 데이터 같이 매우 차원이 높고 희소한 데이터에서는 잘 작동하지 않는데 이 경우엔 선형 모델이 더 적합합니다.

    - 이처럼 랜덤 포레스트는 매우 큰 데이터셋에서 잘 작동하고 여러 코어를 사용해 쉽게 병렬화 할 수 있습니다.
    
    - 하지만 선형 모델보다 메모리 사용량이 많고 훈련과 예측이 느립니다.

- 중요 매개변수는 n_estimators, max_features이고 max_depth와 같은 사전 가지치기 옵션이 있으며 단일 결정 트리의 다른 매개변수도 있습니다.

    - n_estimators는 클수록 더 많은 트리를 평균해 과대적합이 준 안정적인 모델을 만들지만 더 많은 메모리 소모와 더 긴 훈련 시간으로 이어집니다.
    
    - max_features는 각 트리가 얼마나 무작위가 될 것인가를 결정하며 작을수록 과대적합을 줄여줍니다. 가장 좋은 값은 기본값을 쓰는 방식입니다.
    
        - 분류는 n_features ** 0.5, 회귀는 n_features를 max_features로 사용합니다.
    
    - max_features나 max_leaf_nodes를 추가하면 성능이 향상되거나 메모리 소모, 시간을 줄일 수 있습니다.

#### 그레이디언트 부스팅 회귀 트리

- 그레이디언트 부스팅 회귀 트리는 여러 개의 결정 트리를 묶어 강력한 모델을 만드는 또 다른 앙상블 방법입니다.

    - 이름은 회귀이지만 회귀와 분류 모두 사용이 가능합니다.

- 랜덤 포레스트와 달리 그레이디언트 부스팅은 이전 트리의 오차를 보완하는 방식으로 순차적으로 트리를 만듭니다.

    - 기본적으로 그레이디언트 부스팅 회귀 트리는 무작위성이 없으며 대신 강력한 사전 가지치기가 사용됩니다.
    
    - 또한, 보통 하나에서 다섯 정도의 깊지 않은 트리를 사용하므로 메모리를 적게 사용하고 예측도 빠릅니다.

- 그레이디언트 부스팅의 근본 아이디어는 얕은 트리 같은 간단한 모델(**약한 학습기**)을 많이 연결하는 것으로 각 트리는 데이터 일부에 대해서만 예측을 잘 수행하고 트리가 많아질수록 성능이 좋아집니다.

- 그레이디언트 부스팅 트리는 널리 사용되는 알고리즘으로 랜덤 포레스트보다 매개변수 설정에 조금 더 민감하지만 잘 조정하면 더 높은 정확도를 제공합니다.

- 앙상블 방식에 있는 사전 가지치기나 트리 개수 외에도 그레이디언트 부스팅에서 중요한 매개변수는 이전 트리의 오차를 얼마나 강력하게 보정할 지를 제어하는 learning_rate입니다.

    - 학습률이 크면 보정이 강하게 걸려 복잡한 모델을 만듭니다.
    
    - 또한 n_estimators를 키우면 앙상블에 트리가 더 많이 추가되어 모델의 복잡도가 커지고 훈련 세트의 실수를 바로잡을 기회가 많아집니다.

> 다음의 유방암 데이터셋을 가지고 이를 연습해보겠습니다.

```python
from sklearn.ensemble import GradientBoostingClassifier

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train, y_train)

print(f'훈련 세트 정확도: {gbrt.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {gbrt.score(X_test, y_test):.3f}')
```

```python
# 위의 결과는 훈련 세트에 대해 과대적합된 것 같으므로 이를 막기 위해 다음 방법들을 사용하겠습니다.
# 최대 깊이 제한
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)

print(f'훈련 세트 정확도: {gbrt.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {gbrt.score(X_test, y_test):.3f}')

# 학습률 저하
gbrt = GradientBoostingClassifier(random_state=0, learning_rate=0.01)
gbrt.fit(X_train, y_train)

print(f'훈련 세트 정확도: {gbrt.score(X_train, y_train):.3f}')
print(f'테스트 세트 정확도: {gbrt.score(X_test, y_test):.3f}')
```

- 이처럼 모델의 복잡도는 낮춰 훈련 세트의 정확도가 낮아졌습니다.

    - 학습률 저하 방식은 테스트 세트의 정확도가 조금만 상승했지만 최대 깊이를 낮추는 것은 정확도가 꽤나 상승했습니다.

- 다른 결정 트리 기반 모델처럼 모델 특성의 중요도를 시각화하면 모델을 더 잘 이해할 수 있습니다.

```python
gbrt = GradientBoostingClassifier(random_state=0, max_depth=1)
gbrt.fit(X_train, y_train)

plot_feature_importances_cancer(gbrt)
```

- 그레이디언트 부스팅 트리 역시 랜덤 포레스트와 비슷한 특성을 강조하고 있지만 일부 특성을 완전히 무시하고 있습니다.

- 비슷한 종류의 데이터에서 그레이디언트 부스팅과 랜덤 포레스트 둘 다 잘 작동하지만 보통 더 안정적인 랜덤 포레스트를 먼저 적용합니다.

    - 랜덤 포레스트가 잘 작동하더라고 예측 시간이 중요하거나 머신러닝 모델의 마지막 성능까지 쥐어짤 때 그레이디언트 부스팅이 도움이 됩니다.

- 대규모 머신러닝 문제에서 그레이디언트 부스팅을 적용하려면 xgboost 패키지와 python 인터페이스를 검토하는 것이 좋습니다.

#### 장단점과 매개변수

- 그레이디언트 부스팅 결정 트리는 지도 학습에서 가장 강력하고 널리 사용되는 모델 중 하나입니다.

    - 가장 큰 단점은 매개변수를 잘 조정해야 한다는 것과 훈련시간이 길다는 것입니다.
    
        - 또한, 트리 기반 모델의 특성상 희소한 고차원 데이터에는 잘 작동하지 않습니다.

    - 다른 트리 기반 모델처럼 특성의 스케일을 조정하지 않아도 되고 이진 특성이나 연속적인 특성에서도 잘 동작하는 건 장점입니다.

- 그레이디언트 부스팅 트리 모델의 중요 매개변수는 트리의 개수를 지정하는 n_estimators와 이전 트리의 오차를 보정하는 정도를 결정하는 learning_rate 입니다.

    - 이 두 매개변수는 매우 깊게 연관되어 learning_rate를 낮추면 n_estimators를 키워야합니다.
    
    - 하지만 n_estimators가 커지면 모델이 복잡해지고 과대적합될 가능성이 높아집니다.
    
    - 즉, 가용 시간과 메모리 한도 내에서 적절한 n_estimators를 맞추고 learning_rate를 맞추는 것이 좋습니다.

        - scikit-learn에서는 조기 종료를 위한 n_iter_no_change와 validation_fraction이 추가되었는데 전자는 후자의 데이터를 사용해 해당 반복 동안 검증 점수가 향상되지 않으면 훈련 종료를, 후자는 해당 배율 만큼 검증 데이터를 사용한다는 의미입니다.

- 중요한 또 다른 매개변수는 각 트리의 복잡도를 낮추는 max_depth 또는 max_leaf_nodes로 통상 그레이디언트 부스팅 모델은 max_depth를 매우 작게 설정하여 트리의 깊이가 5보다 깊어지지 않도록 합니다.

### 2.3.7 (+plus) 배깅, 엑스트라 트리, 에이다부스트

- 앞서 대표적인 앙상블 방법인 랜덤 포레스트와 그레이디언트 부스팅 알고리즘을 살펴보았다면 이번 절은 scikit-learn의 다른 앙상블 알고리즘은 배깅, 에이다부스트, 엑스트라 트리에 대해 알아보겠습니다.

- 여기에서도 동일한 two_moons 데이터셋과 cancer 데이터셋을 사용하겠으니 이들을 먼저 import 하겠습니다.

```python
from preamble import *
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons, load_breast_cancer

Xm, ym = make_moons(n_samples=100, noise=0.25, random_state=3)
Xm_train, Xm_test, ym_train, ym_test = train_test_split(Xm, ym, stratify=ym, random_state=42)

cancer = load_breast_cancer()
Xc_train, Xc_test, yc_train, yc_test = train_test_split(cancer.data, cancer.target, random_state=0)
```

#### 배깅

- **배깅**은 Bootstrap aggregating의 줄임말로 중복을 허용한 랜덤 샘플링으로 만든 훈련 세트를 사용해 분류기를 각각 다르게 학습시킵니다.

    - 부트스트랩 샘플은 앞서 살펴본 랜덤 포레스트와 동일하게 만들고 predict_proba() 메서드를 사용해 확률 값을 평균하여 예측을 수행합니다.
    
    - 해당 메서드가 없는 경우엔 빈도가 가장 높은 클래스 레이블이 예측 결과가 됩니다.

- 배깅을 통해 cancer 데이터셋에 로지스틱 회귀 모델 100개를 훈련해 앙상블하면 다음과 같습니다.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier

bagging = BaggingClassifier(LogisticRegression(), n_estimators=100, oob_score=True, n_jobs=-1, random_state=42)
bagging.fit(Xc_train, yc_train)
```

- 위 코드에서 LogisticRegression 객체를 기반 분류기로 전달하고 100개를 훈련하도록 지정했습니다.

    - oob_score를 True로 지정하면 매개변수는 부트스트래핑에 포함되지 않은 샘플을 기반으로 훈련 모델을 평가하고 이를 통해 테스트 세트의 성능을 짐작할 수 있습니다.
    
    - RandomForestClassifier도 해당 매개변수를 지원하며 둘 다 기본 값은 False입니다.

```python
print(f'훈련 세트 정확도: {bagging.score(Xc_train, yc_train):.3f}')
print(f'테스트 세트 정확도: {bagging.score(Xc_test, yc_test):.3f}')
print(f'OOB sample 정확도: {bagging.oob_score_:.3f}')
```

```python
# 결정 트리로 배깅을 하는 것보다 랜덤 포레스트가 더 편리하지만 여기저는 결정 트리에 배깅을 해보겠습니다.
from sklearn.tree import DecisionTreeClassifier

bagging = BaggingClassifier(DecisionTreeClassifier(), n_estimators=5, n_jobs=-1, random_state=42)
bagging.fit(Xm_train, ym_train)
```

```python
# 이제 이 결정 트리를 시각화해보겠습니다. 결과는 랜덤 포레스트와 매우 비슷합니다.
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for i, (ax, tree) in enumerate(zip(axes.ravel(), bagging.estimators_)):
    ax.set_title(f'tree {i}')
    mglearn.plots.plot_tree_partition(Xm, ym, tree, ax=ax)
    
mglearn.plots.plot_2d_separator(bagging, Xm, fill=True, ax=axes[-1, -1], alpha=0.4)
axes[-1, -1].set_title('bagging')
mglearn.discrete_scatter(Xm[:, 0], Xm[:, 1], ym)
plt.show()
```

```python
# 이번에는 n_estimators=100으로 늘리고 cancer에 대해 훈련한 뒤 훈련 세트와 테스트 세트의 성능을 보겠습니다.
bagging = BaggingClassifier(DecisionTreeClassifier(), n_estimators=100, oob_score=True, n_jobs=-1, random_state=42)
bagging.fit(Xc_train, yc_train)

print(f'훈련 세트 정확도: {bagging.score(Xc_train, yc_train):.3f}')
print(f'테스트 세트 정확도: {bagging.score(Xc_test, yc_test):.3f}')
print(f'OOB sample 정확도: {bagging.oob_score_:.3f}')
```

- 배깅은 랜덤 포레스트와 달리 max_samples 매개변수에서 부트스트랩 샘플의 크기를 지정할 수 있습니다.

- 또한 랜덤 포레스트는 DecisionTreeClassifier(splitter='best')를 사용하도록 고정되어 있습니다.

    - 여기서 결정 트리의 splitter='random'으로 설정하면 무작위로 분할한 후보 노드 중 최선의 분할을 찾습니다.

- 이런 결정 트리의 배깅 앙상블은 다음에 설명할 엑스트라 트리와 관련이 많습니다.
