    - 따라서 비지도학습의 학습 알고리즘은 입력 데이터만으로 데이터에서 지식을 추출할 수 있어야 합니다.

## 3.1 비지도 학습의 종류

- 이번 장에서는 두 가지 비지도 학습을 살펴보겠습니다. 바로 데이터의 비지도 변환과 군집입니다.

- **비지도 변환**은 데이터를 새롭게 표현하여 사람이나 머신러닝 알고리즘이 원래 데이터보다 쉽게 해석할 수 있도록 만드는 알고리즘입니다.

    - 비지도 변환이 널리 사용되는 분야는 많은 고차원 데이터의 특성 수를 줄이면서 꼭 필요한 특징만 포함한 데이터로 표현하는 차원 축소로 대표적으로 시각화하기 위해 데이터셋을 2차원으로 변경하는 경우입니다.
    
    - 비지도 변환에서 데이터를 구성하는 단위나 성분을 찾기도 하는데 많은 텍스트 문서에서 주제를 추출하는 것이 이런 예시 입니다. 이 때 처리할 작업은 문서에서 이야기하는 주제들이 무엇인지 찾고 학습하는 것입니다.

- **군집 알고리즘**은 데이터를 비슷한 것끼리 그룹으로 묶는 것으로 대표적으로 소셜 미디어 사이트에서 업로드한 사진을 자동으로 분류하는 것입니다.

## 3.2 비지도 학습의 도전 과제

- 비지도 학습에서 가장 어려운 일은 알고리즘이 뭔가 유용한 것을 학습했는지 평가하는 것입니다.

    - 비지도 학습은 보통 레이블이 없는 데이터에 적용하기 때문에 무엇이 올바른 출력인지 모릅니다. 그래서 어떤 모델이 일을 잘하고 있는지 판단하기가 매우 어렵습니다.
    
    - 더욱이 비지도 학습 모델에 우리가 원하는 것을 알려줄 방법이 없기 때문에 비지도 학습의 결과를 평가하기 위해서는 직접 확인하는 것이 유일한 방법일 때가 많습니다.

- 이런 이유로 비지도 학습 알고리즘은 데이터 과학자가 데이터를 더 잘 이해하고 싶을 때 탐색적 분석 단계에서 많이 사용하며 지도 학습의 전처리 단계에서도 많이 사용합니다.

    - 비지도 학습의 결과로 새롭게 표현된 데이터를 사용해 학습하면 지도 학습의 정확도가 좋아지기도 하고 메모리와 시간을 절약할 수 있습니다.

- 진짜 비지도 학습 알고리즘으로 들어가기 전에 몇 가지 간단한 전처리 메서드를 짧게 설명하겠습니다. 지도 학습 알고리즘에서 전처리와 스케일 조정을 자주 사용하지만, 스케일 조정 메서드는 지도 정보를 사용하지 않으므로 비지도 방식입니다.

- 앞 장에서 본 신경망과 SVM 같은 알고리즘은 데이터의 스케일에 매우 민감합니다. 그래서 보통 이런 알고리즘에 맞게 데이터의 특성 값을 조정합니다.

    - 보통 특성마다 스케일을 조정해서 데이터를 변경합니다.

```python
import mglearn

mglearn.plots.plot_scaling()
```

### 3.3.1 여러 가지 전처리 방법

- 앞 선 그림의 첫 번째 그래프는 두 개의 특성을 인위적으로 만든 이진 분류 데이터셋으로 x축 10 ~ 15, y축 1 ~ 9 사이에 데이터가 존재합니다.

- 그 다음 오른쪽 네 개의 그래프는 이들을 기준이 되는 범위로 변환하는 방법들로 scikit-learn의 StandardScaler는 각 특성의 평균을 0, 분산을 1로 변경하여 모든 특성이 같은 크기를 가지게 합니다.

- 그러나 이 방법은 특성의 최솟값과 최댓값을 제한하지 않는데 RobustScaler는 특성들이 같은 스케일을 갖게 하는 점에서 StandardScaler와 비슷하지만 중간값과 사분위값을 사용해 이를 처리합니다.
    
    - 이런 방식 덕분에 RobustScaler는 전체 데이터와 아주 동떨어진 데이터 포인트인 **이상차**에 영향을 받지 않습니다.
    
- 반면에 MinMaxScaler는 모든 특성이 정확하게 0과 1사이에 위치하도록 데이터를 변경하며 모든 데이터가 x, y 값이 0과 1사이로 조정됩니다.

- 마지막으로 Normalizer는 매우 다른 스케일 조정 기법으로 특성 벡터의 유클라디안 길이가 1이 되도록 데이터 포인트를 조정합니다.

    - 다시 말해 각 데이터 포인트가 다른 비율로 길이에 반비례해 스케일이 조정된다는 뜻으로 정규화는 특성 벡터의 길이는 상관없고 데이터의 방향 혹은 각도가 중요할 때 많이 사용합니다.

### 3.3.2 데이터 변환 적용하기

- 여러 종류의 변환을 scikit-learn을 사용해 실제로 적용해보겠습니다. 데이터셋은 2장에서 사용한 cancer를 사용합니다.

    - 스케일을 조정하는 전처리 메서드들은 보통 지도 학습 알고리즘을 적용하기 전에 적용하며 여기서는 cancer에 커널 SVM(SVC)를 적용하고 전처리에 MinMaxScaler를 사용하겠습니다.

- 먼저 데이터를 적재하고 훈련 세트와 테스트 세트로 나눕니다.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)

print(X_train.shape, X_test.shape)
```

- 이 데이터셋에는 569개의 데이터 포인트가 있고 각 데이터 포인트는 30개의 측정값으로 이뤄져 있습니다. 여기서는 426개를 훈련 세트로, 143개를 테스트 세트로 나누었습니다.

- 이전에 만든 지도 학습 모델처럼 먼저 전처리가 구현된 python 클래스를 import하고 객체를 생성합니다.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

- 그런 다음 fit 메서드를 훈련 데이터에 적용하는데 MinMaxScaler의 fit은 훈련 세트의 특성마다 최솟값과 최댓값을 계산합니다. 2장의 분류 모델이나 회귀 모델과는 달리 스케일 객체는 훈련 데이터만 사용하며 결과 레이블은 사용하지 않습니다.

```python
scaler.fit(X_train)
```

- fit 메서드로 학습한 변환을 적용하려면, 즉 실제 데이터의 스케일을 조정하려면 스케일 객체의 transform 메서드를 사용합니다. 이는 새로운 데이터 표현을 만들 때 사용합니다.

```python
# 데이터 변환 과정입니다.
X_train_scaled = scaler.transform(X_train)

# 결과를 확인합니다.
print('변환 후 크기:', X_train_scaled.shape)
print('스케일 변환 전 특성별 최소값:\n', X_train.min(axis=0))
print('스케일 변환 전 특성별 최대값:\n', X_train.max(axis=0))
print('스케일 변환 후 특성별 최소값:\n', X_train_scaled.min(axis=0))
print('스케일 변환 후 특성별 최대값:\n', X_train_scaled.max(axis=0))
```

- 변환된 데이터의 배열 크기는 원래 데이터와 동일하므로 특성 값이 이동되거나 크기가 조정되기만 했음을 보입니다. 예상대로 모든 특성 값은 0과 1 사이가 되었습니다.

- 이 데이터에 SVM을 적용하려면 테스트 세트도 변환해야 합니다. 이 땐 X_test를 transform 메서드에 넣고 활용합니다.

```python
# 데이터 변환 과정입니다.
X_test_scaled = scaler.transform(X_test)

# 결과를 확인합니다.
print('스케일 변환 후 특성별 최소값:\n', X_test_scaled.min(axis=0))
print('스케일 변환 후 특성별 최대값:\n', X_test_scaled.max(axis=0))
```

- 보시다시피 스케일을 조정한 테스트 세트의 최솟값과 최댓값이 0과 1이 아니며 일부 특성은 범위를 벗어났습니다.

- 이는 MinMaxScaler와 다른 모든 스케일 모델도 항상 훈련 세트와 테스트 세트에 같은 변환을 적용하기 때문입니다.

    - transform 메서드는 테스트 세트의 최솟값과 최댓값의 범위가 아닌 훈련 테스트의 최솟값을 뺀 뒤 훈련 세트의 범위로 나눕니다.

### 3.3.3 QuantileTransformer와 Power Transformer

- scikit-learn 0.19.0 이상부터 QuantileTransformer가 추가되었습니다. 해당 기능은 기본 1000개의 분위를 사용해 데이터를 균등하게 분포시킵니다. RobustScaler와 비슷하게 이상치에 민감하지 않으며 전체 데이터를 0과 1사이로 압축합니다.

- QuantileTransformer를 사용해보겠습니다. 예시는 다음과 같습니다.

```python
# 우신 필요한 module을 import 합니다.
from sklearn.preprocessing import QuantileTransformer, StandardScaler, PowerTransformer
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 이전과 동일한 예제 데이터를 생성하고 산점도를 그리겠습니다.
X, y = make_blobs(n_samples=50, centers=2, random_state=4, cluster_std=1)
X += 3

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, edgecolors='black')
plt.xlim(0, 16)
plt.xlabel('x0')
plt.ylim(0, 10)
plt.ylabel('x1')
plt.title('Original Data')
plt.show()
```

```python
# 이제 이 데이터셋에 QuantileTransformer를 적용해 스케일의 변화를 확인하겠습니다.
scaler = QuantileTransformer()
X_trans = scaler.fit_transform(X)

plt.scatter(X_trans[:, 0], X_trans[:, 1], c=y, s=30, edgecolors='black')
plt.xlim(0, 5)
plt.xlabel('x0')
plt.ylim(0, 5)
plt.ylabel('x1')
plt.title(type(scaler).__name__)
plt.show()
```

```python
# 랜덤하게 흩어졌던 데이터 포인트들이 0과 1사이에 고르게 분포되었습니다. 이를 히스토그램을 통해 더 자세히 알아보겠습니다.
plt.hist(X_trans)
plt.show()
```

- QuantileTransformer의 분위 수는 n_quantiles 매개변수에서 설정할 수 있고 기본 값은 1000입니다.

    - scaler 객체의 quantiles_ 속성에 특성별로 계산된 분위 값이 들어있고 이 크기는 (n_quantiles, n_features)입니다. 이를 확인하면 다음과 같습니다.

```python
print(scaler.quantiles_.shape)
```

- QuantileTransformer의 동작을 조금 더 잘 이해하기 위해 간단한 샘플 데이터를 만들어 변환 결과를 비교해보겠습니다.

    - 아래 코드는 하나의 특성과 다섯 개의 샘플로 이루어진 간단한 데이터셋으로 np.percentile() 함수는 두 번째 매개변수에 지정한 분위에 해당하는 샘플을 추출합니다.
    
        - 본 예시에서는 첫 번째, 1사분위, 중간값, 3사분위, 마지막 값으로 지정했습니다.

    - 훈련 샘플이 모두 다섯 개 이므로 지정한 분위가 훈련 샘플과 정확히 일치합니다.

```python
x = np.array([[0], [5], [8], [9], [10]])
print(np.percentile(x[:, 0], [0, 25, 50, 75, 100]))
```

- 이를 QuantileTransformer로 변환한 후 확인해부겠습니다. 다음 결과를 통해 예제 데이터가 분위가 동일한 값으로 변환됨을 알 수 있습니다.

```python
x_trans = QuantileTransformer().fit_transform(x)
print(np.percentile(x_trans[:, 0], [0, 25, 50, 75, 100]))
```

- QuantileTransformer는 output_distribution 매개변수를 normal로 지정하여 균등분포가 아닌 정규분포로 출력할 수 있습니다. 그 경우는 다음과 같습니다.

```python
scaler = QuantileTransformer(output_distribution='normal')
X_trans = scaler.fit_transform(X)

plt.scatter(X_trans[:, 0], X_trans[:, 1], c=y, s=30, edgecolors='black')
plt.xlim(-5, 5)
plt.xlabel('x0')
plt.ylim(-5, 5)
plt.ylabel('x1')
plt.title(type(scaler).__name__)
plt.show()
```

- scikit-learn 0.20.0 이상에서는 데이터 특성별로 정규분포 형태에 가깝도록 변환해주는 PowerTransformer가 추가되었습니다.

    - PowerTransformer는 method 매개변수에 'yeo-johnson'과 'box-cox' 알고리즘을 지정할 수 있으며 기본 값은 전자입니다.
    
- QuantileTransformer, StandardScaler, PowerTransformer의 알고리즘들이 예제 데이터셋을 정규분포로 변환하는 차이를 확인해보겠습니다.

```python
plt.hist(X)
plt.title('Original Data')
plt.show()

X_trans = QuantileTransformer(output_distribution='normal').fit_transform(X)
plt.hist(X_trans)
plt.title('QuantileTransformer')
plt.show()

X_trans = StandardScaler().fit_transform(X)
plt.hist(X_trans)
plt.title('StandardScaler')
plt.show()

X_trans = PowerTransformer(method='box-cox').fit_transform(X)
plt.hist(X_trans)
plt.title('PowerTransformer box-cox')
plt.show()

X_trans = PowerTransformer(method='yeo-johnson').fit_transform(X)
plt.hist(X_trans)
plt.title('PowerTransformer yeo-johnson')
plt.show()
```

- 이 예에서는 PowerTransform의 두 알고리즘이 동일한 결과를 만들었지만 실전에서는 데이터셋마다 어떤 변환이 정규분포에 가깝게 변환될지 사전에 알기 어렵습니다. 따라서 각 변환기의 결과를 히스토그램으로 확인하는 것이 좋습니다.

### 3.3.4 훈련 데이터와 테스트 데이터의 스케일을 같은 방법으로 조정하기

- 지도 학습 모델에서 테스트 세트를 사용하려면 훈련 세트와 테스트 세트에 같은 변환을 적용해야 한다는 점이 중요합니다.

- 다음 예는 테스트 세트의 최솟값과 범위를 사용하면 어떤 일이 일어나는 지 알 수 있습니다.

```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
import mglearn

# 인위적 데이터를 생성하고 훈련 세트와 테스트 세트로 나눕니다.
X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)

# 훈련 세트와 테스트 세트의 산점도를 그립니다.
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:, 0], X_train[:, 1], c=mglearn.cm2.colors[0], label='train set', s=60)
axes[0].scatter(X_test[:, 0], X_test[:, 1], marker='^', c=mglearn.cm2.colors[1], label='test set', s=60)
axes[0].legend(loc='upper left')
axes[0].set_title('Original Data')

# MinMaxScaler를 사용해 스케일을 조정합니다.
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 스케일이 조정된 데이터의 산점도를 그립니다.
axes[1].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=mglearn.cm2.colors[0], label='train set', s=60)
axes[1].scatter(X_test_scaled[:, 0], X_test_scaled[:, 1], marker='^', c=mglearn.cm2.colors[1], label='test set', s=60)
axes[1].set_title('Scale Adjusted Data')

# 테스트 세트의 스케일을 따로 조정하되 최솟값은 0, 최댓값이 1이 되도록 합니다. 이는 확인용으로 실전에서는 금물입니다.
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)

# 잘못 조정된 데이터의 산점도를 그립니다.
axes[2].scatter(X_train_scaled[:, 0], X_train_scaled[:, 1], c=mglearn.cm2.colors[0], label='train set', s=60)
axes[2].scatter(X_test_scaled_badly[:, 0], X_test_scaled_badly[:, 1], marker='^', c=mglearn.cm2.colors[1], label='test set', s=60)
axes[2].set_title('Scale Not Adjusted Data')

for ax in axes:
    ax.set_xlabel('attribute 0')
    ax.set_ylabel('attribute 1')
```

- 첫 번째 그래프는 2차원 원본 데이터셋으로 훈련 세트는 원, 테스트 세트는 삼각형입니다. 두 번째 그래프는 MinMaxScaler로 스케일을 조정했으며 훈련 세트를 사용해 fit 메서드를 호출하고 훈련 세트와 테스트 세트에 transform 메서드를 적용했습니다.

    - 이 둘은 축의 눈금이 바뀐 것을 제외하고 동일하며 모든 특성이 0과 1 사이에 위치하게 변했습니다. 물론 테스트 데이터의 최솟값은 0이 아닙니다.

- 세 번째 그래프는 훈련 세트와 테스트 세트의 스케일을 서로 다른 방식으로 조정한 것으로 두 세트 모두 최솟값과 최댓값이 0과 1 사이지만 데이터셋이 매우 다르게 보입니다.

    - 테스트 포인트가 스케일이 다르게 조정되어 훈련 세트와 다른 비율로 이동했습니다. 따라서 데이터 배열이 뒤죽박죽되었습니다.

- 다르게 보면 포인트가 하나뿐인 테스트 세트에서 하나의 포인트로는 MinMaxScaler에 필요한 최솟값과 최댓값이 없어 스케일을 조정할 수 없습니다. 하지만 테스트 세트의 크기 때문에 작업 과정이 달라져서는 안됩니다.

#### 단축 메서드와 효율적인 방법

- 보통 어떤 데이터셋에 fit을 적용하면 transform을 호출합니다. 이는 매우 전형적인 방법으로 이 둘을 차례로 호출하는 것보다 효율적인 방법이 필요합니다.

- 이럴 때를 위해 transform 메서드를 가진 모델은 fit_transform 메서드를 제공하며 그 사용 예시는 다음과 같습니다.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# 메소드 chaning을 사용해 fit과 transform을 연달아 호출합니다.
X_scaled = scaler.fit(X_train).transform(X_train)
# 이를 더 쉽게 사용하는 방법은 다음과 같습니다.
X_scaled_new = scaler.fit_transform(X_train)
```

### 3.3.5 지도 학습에서 데이터 전처리 효과

- 이제 다시 cancer 데이터셋으로 돌아가 SVC를 학습시킬 때 MinMaxScaler의 효과를 확인해보겠습니다. 우선 원본 데이터로 SVC 모델을 만듭니다.

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

svm = SVC(C=100)
svm.fit(X_train, y_train)
print(f'테스트 세트 정확도: {svm.score(X_test, y_test):.2f}')
```

- 이제 SVC 모델을 학습시키기 전에 MinMaxScaler를 사용해 데이터의 스케일을 조정합니다.

```python
# 0과 1 사이로 스케일을 조정합니다.
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 조정된 데이터로 학습한 후 정확도를 확인합니다.
svm.fit(X_train_scaled, y_train)
print(f'테스트 세트 정확도: {svm.score(X_test_scaled, y_test):.2f}')
```

- 앞서 본 것처럼 스케일 조정의 효과는 큰 편으로 데이터 스케일 조정에 복잡한 수학이 필요하진 않지만, 실수를 줄이기 위해 직접 구현하기보다 scikit-learn에서 제공하는 도구를 사용하는 것이 좋습니다.

- 모든 전처리 모델이 동일한 fit, transform 메서드를 제공하므로 사용하는 python class만 바꾸면 다른 전처리 알고리즘으로 손쉽게 교체할 수 있습니다.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm.fit(X_train_scaled, y_train)
print(f'테스트 세트 정확도: {svm.score(X_test_scaled, y_test):.2f}')
```

### 3.4.1 주성분 분석(PCA)

- 주성분 분석은 특성들이 통계적으로 상관관계가 없도록 데이터셋을 회전시키는 기술로 회전한 뒤 데이터를 설명하는 데 얼마나 중요하냐에 따라 종종 새로운 특성 중 일부만 선택됩니다.

    - 다음 예제는 인위적으로 만든 2차원 데이터셋을 사용해 PCA 효과를 나타낸 것입니다.

```python
mglearn.plots.plot_pca_illustration()
```

- 첫 번째 그래프는 원본 데이터 포인트를 색으로 구분해 표시한 것으로 이 알고리즘은 먼저 '성분 1'이라고 쓰인 분산이 가장 큰 방향을 찾습니다.

    - 다른 말로 특성들의 상관관계가 가장 큰 방향을 찾습니다.

- 다음으로 첫 번째 방향과 직각인 방향 중 가장 많은 정보를 담은 방향을 찾습니다.

    - 2차원에서는 가능한 직각 방향이 하나뿐이지만 고차원에서는 무한히 많은 직각 방향이 있을 수 있습니다.

    - 여기에선 두 성분을 화살표로 나타내었지만 화살표의 머리와 꼬리는 아무 의미가 없습니다. 즉, 화살표를 중앙에서 오른쪽 아래가 아닌 중앙에서 왼쪽 위로 향하게 그릴 수도 있습니다.

- 이런 과정을 거쳐 찾은 방향을 데이터에 있는 주된 분산의 방향이라고 하여 **주성분**이라고 합니다. 일반적으로 원본 특성 개수만큼 주성분이 있습니다.

- 두 번째 그래프는 같은 데이터의 주성분 1과 2를 각각 x, y축에 나란하도록 회전한 것으로 회전하기 전에 데이터의 평균을 빼 중심을 원점에 맞추었습니다.

    - PCA에 의해 회전된 두 축은 연관되어 있지 않아 변환된 데이터의 상관관계 행렬이 대각선 방향을 제외하고 0이 됩니다.

- PCA는 주성분의 일부만 남기는 차원 축소 용도로 사용할 수 있는데 그 예시는 세 번째 그래프와 같습니다. 이런 과정을 거치면 2차원 데이터셋이 1차원으로 차원이 감소합니다.

    - 그러나 단순히 원본 특성 중 하나만 남기는 것이 아닌 가장 유용한 방향을 찾아서 그 방향의 성분, 즉 첫 번째 주성분을 유지합니다.

- 마지막으로 데이터에 다시 평균을 더하고 반대로 회전하면 마지막 그래프가 나오는데 이 때 데이터 포인트들이 원래 특성 공간에 있지만 첫 번째 주성분의 정보만 담고 있습니다.

- 이 변환은 데이터에서 노이즈를 제거하거나 주성분에서 유지되는 정보를 시각화하는 데 종종 사용합니다.
