## 1.1 왜 머신러닝인가?

- 초창기 지능형 애플리케이션들은 `if`와 `for`의 향연으로 하드 코딩되어 있었습니다.

- 이런 방식을 **규칙 기반 전문가 시스템**이라고 하지만 이를 직접 구현하는 건 큰 단점이 있습니다.

    1. 결정에 필요한 로직이 특정 한 분야나 작업에 국한됩니다. 조금만 변경되어도 대대적인 수정이 필요합니다.
    
    2. 해당 분야 전문가들이 내리는 결정 방식에 대해 알아야 규칙을 설계할 수 있습니다.

- 이 문제를 해결하기 위해 머신러닝을 사용하게 되었습니다.

### 1.1.1 머신러닝으로 풀 수 있는 문제

- 가장 많이 사용되는 머신러닝 알고리즘은 이미 알려진 사례를 바탕으로 일반화된 모델을 만드는 **지도 학습**입니다.

    - 지도 학습은 사용자가 알고리즘에 입력과 그에 상응하는 출력을 제공하고 이를 알고리즘이 적절한 출력을 만드는 것입니다.

- 이렇게 학습된 알고리즘은 사람의 도움 없이도 새로운 입력을 받으면 적절한 출력을 만들 수 있습니다.

- 지도 학습 알고리즘은 분석하기 좋고 성능을 측정하기 쉽지만 입출력 데이터를 수작업으로 만들어야 합니다.

    - 지도 학습의 예로 우편번호 분석, 의료 영상을 기반으로 한 병 판단, 신용카드 부정 거래 감지 등이 있습니다.

- 반면 알고리즘에 입력은 주어지나 출력이 주어지지 않는 형태의 학습도 존재합니다. 이를 **비지도 학습**이라고 합니다.

    - 비지도 학습의 예로 블로그 글의 주제 구분, 고객의 취향 분석 및 그룹화, 비정상적 웹사이트 접근 탐지 등이 있습니다.

- 이들 학습은 모두 컴퓨터가 인식 가능한 형태의 데이터가 필요합니다.

    - 이 데이터 중 하나의 개체 혹은 행을 **샘플** 혹은 **데이터 포인트**라고 합니다.
    
    - 반면 샘플의 특성, 즉 열을 **특성**이라고 합니다.
    
    - 이들을 통해 좋은 입력 데이터를 만드는 것을 **특성 추출** 혹은 **특성 공학**이라고 합니다.

### 1.4.3 SciPy

- [SciPy](https://www.scipy.org/scipylib)는 과학 계산용 함수를 모아놓은 python 패키지입니다.

- SciPy에는 고성능 선형 대수, 함수 최적화, 신호처리, 특수 수학 함수와 통계 분포 등 다양한 기능을 제공합니다.

- scikit-learn의 주요 알고리즘을 구현할 때 SciPy를 사용합니다.

    - 그 중 핵심 기능은 `scipy.sparse`로 scikit-learn에서 데이터를 표현하는 방법 중 하나인 희소 행렬 기능입니다.
    
    > 다음은 SciPy 코드의 예제입니다.

````python
from scipy import sparse

# 대각선 원소는 1, 나머지는 0인 2차원 NumPy 배열입니다.
eye = np.eye(4)
print('NumPy:\n', eye)

# NumPy 배열을 CSR 포맷의 SciPy 희박 행렬로 변환합니다.(즉, 0이 아닌 원소만 저장됩니다.)
sparse_matrix = sparse.csr_matrix(eye)
print('SciPy CSR.ver:\n', sparse_matrix)
````

- 보통 희소 행렬을 0이 모두 채워진 2차원 배열로부터 만들지 않으므로 희소 행렬을 직접 만들 수 있어야 합니다.

> 다음은 COO 포맷을 사용해 이전과 동일한 희소 행렬을 만드는 예제입니다.

````python
# 우선 NumPy 배열을 만듭니다.
data = np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)

# COO 변환
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print('SciPy COO.ver:\n', eye_coo)
````

### 1.4.5 pandas

````python
from IPython.display import display
import pandas as pd

# 간단한 데이터셋(회원정보)를 생성합니다.
data = {
    'Name': ['J', 'K', 'L', 'P'],
    'Location': ['Seoul', 'Daejeon', 'Gwangju', 'Busan'],
    'Age': [24, 26, 25, 28]
}

# IPython을 사용해 pandas 데이터를 예쁘게 출력합니다.
data_pandas = pd.DataFrame(data)
display(data_pandas)

# Age가 26 이상인 모든 행을 출력합니다.
display(data_pandas[data_pandas.Age >= 26])
````

## 1.7 첫 번째 애플리케이션: 붓꽃의 품종 분류

- 이번 장에서 간단한 머신러닝 애플리케이션을 훑어가며 첫 번째 모델을 만들어보겠습니다.

- 조건은 다음과 같습니다.

    - 붓꽃의 품종을 알기위해 붓꽃의 꽃잎과 꽃받침의 폭과 길이를 cm 단위로 측정한 데이터가 있습니다.
    
    - 또, 붓꽃을 setosa, versicolor, virginica 종의 붓꽃 측정 데이터도 있습니다.
    
    - 위 데이터들을 통해 붓꽃을 분류하는 머신러닝 모델을 만들어보세요.

- 붓꽃의 품종을 정확하게 분류한 데이터가 있으므로 이 문제는 **지도 학습**의 한 종류입니다.

- 또한, 몇 가지 선택지 중 하나를 선택하므로 **분류** 문제입니다.

    - 여기서 출력 가능한 모든 값을 **클래스**라고 합니다.

- 따라서, 세 개의 클래스를 분류하는 문제입니다.

- 데이터 포인트 하나에 대한 기대 출력은 꽃의 품종입니다. 이런 데이터의 출력을 **레이블**이라고 합니다.

### 1.7.1 데이터 적재

- 사용할 데이터셋은 오래전부터 사용해온 붓꽃 데이터셋입니다.

````python
from sklearn.datasets import load_iris  # 붓꽃 데이터는 이미 scikit-learn 데이터셋에 포함되어 있습니다.

iris_dataset = load_iris()  # 해당 데이터의 형태는 dict와 유사한 bunch class입니다.
print(f'iris_dataset의 키: {iris_dataset.keys()}')
print()

# DESCR에는 데이터셋에 관한 간략한 설명이 들어있습니다.
print(iris_dataset['DESCR'][:193] + '\n...')
print()

# target_names에는 예측하려는 붓꽃 품종의 이름을 문자열 배열로 가지고 있습니다.
print(f'target의 이름: {iris_dataset["target_names"]}')
print()

# feature_names에는 각 특성을 설명하는 문자열 배열을 가지고 있습니다.
print(f'feature의 이름: {iris_dataset["feature_names"]}')
print()

# 실제 데이터는 target과 data에 들어있습니다.
# data는 꽃잎의 길이, 폭, 꽃받침의 길이, 폭을 가진 NumPy 배열입니다.
print(f'data의 타입: {type(iris_dataset["data"])}')
print()

# data 배열의 행을 각각의 꽃이며 열은 각 꽃에서 구한 네 개의 측정치입니다.
# 여기서 배열의 각 아이템을 머신러닝은 `샘플`, 속성은 `특성`으로 부릅니다.
print(f'data의 크기: {iris_dataset["data"].shape}')
print()

# 맨 처음 다섯 샘플의 특성값입니다.
print(f'data의 첫 다섯 행: {iris_dataset["data"][:5]}')
print()

# target 배열도 붓꽃의 품종을 담은 NumPy 배열입니다.
print(f'target의 타입: {type(iris_dataset["target"])}')
print()

# target 배열은 닫 원소가 붓꽃 하나에 해당하는 1차원 배열입니다.
print(f'target의 크기: {iris_dataset["target"].shape}')
print()

# target 배열에는 붓꽃의 종류가 0에서 2까지 숫자로 기록되어 있습니다.
print(f'target: {iris_dataset["target"]}')
print()

# 각 숫자의 의미는 target_names에서 확인할 수 있습니다.
print(f'target number의 의미: {iris_dataset["target_names"]}')
````

### 1.7.2 성과 측정: 훈련 데이터와 테스트 데이터

- 이 데이터로 머신러닝 모델을 만들고 새로운 데이터의 품종을 예측하려 하기 위해서 만든 모델이 진짜 잘 작동하는지 알아야 합니다.

- 하지만 검증할 때 모델을 만드는 데 사용한 데이터를 평가용으로 사용할 수 없습니다.

    - 이는 모델이 훈련 데이터를 전부 기억하여 전부 맞춰버릴 수 있기 때문입니다. 이를 **일반화**하지 않았다고 합니다.

- 따라서, 모델의 성능을 측정하기 위해서 레이블을 알고 있지만 이전에 본 적 없는 새 데이터가 필요합니다.

    - 이를 위해 가지고 있는 데이터를 두 그룹으로 나눕니다.
    
    - 이 중 하나는 머신러닝 모델을 만드는 **훈련 데이터** 혹은 **훈련 세트**입니다.
    
    - 다른 하나는 검증에 사용되는 **테스트 데이터** 혹은 **테스트 세트**, **홀드아웃 세트**라고 합니다.

- scikit-learn은 데이터셋을 섞어서 나눠주는 `train_test_split` 함수를 제공합니다.

    - 이 함수는 전체 행중 75%를 훈련 세트로, 25%를 테스트 세트로 분류합니다.
    
    - 이 비율을 조정할 수 있으나 일반적으로는 이 비율을 사용합니다.

- scikit-learn에서 데이터는 `X`, 레이블은 `y`라고 합니다.

    - 다만 데이터는 2차원 배열이므로 `X`, 레이블은 1차원 배열이므로 `y`로 대소를 구별합니다.

````python
from sklearn.model_selection import train_test_split

# 일반적으로 데이터 그룹화의 변수명은 다음과 같이 사용합니다.
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
````

- `train_test_split` 함수로 데이터를 나누기 전에 유사 난수 생성기로 데이터셋을 무작위로 섞어야 합니다.
    
    - 만일 섞이지 않았다면 이 모델이 일반화 되었는지 검증할 수 없습니다.

- 따라서 `random_state` 인자를 통해서 난수 초깃값을 결정할 수 있습니다.

- 나뉘어진 데이터셋 그룹의 비중은 다음을 통해 확인할 수 있습니다.

````python
print(f'X_train의 크기: {X_train.shape}')
print(f'y_train의 크기: {y_train.shape}')
print(f'X_test의 크기: {X_test.shape}')
print(f'y_test의 크기: {y_test.shape}')
````

### 1.7.3 가장 먼저 할 일: 데이터 살펴보기

- 머신러닝 모델을 만들기 전에 해당 문제가 머신러닝 없이 풀 수 있는 문제는 아닌지, 혹은 필요한 정보가 누락되지 않았는지 확인이 필요합니다.

- 또한, 데이터를 탐색하며 비정상적인 값이나 특이한 값을 찾을 수도 있습니다.

- 이를 해결하기 위해 데이터의 **시각화**는 매우 중요하며 **산점도**를 확인하는 건 매우 좋은 방법 중 하나입니다.

    - 산점도는 데이터의 한 특성을 x, 다른 하나를 y에 놓고 비교하는 그래프입니다.

- 다만, 2차원 혹은 3차원의 데이터 밖에 보지 못하는 단점이 있습니다.

    - 이를 해결하려면 **산점도 행렬**을 사용해서 확인해야 합니다.

    - 주의할 점은 산점도 행렬은 한 그래프에서 모든 특성의 관계가 나오는 것이 아니기에 그래프에 드러나지 않은 중요한 성질이 있을 수 있습니다.

- 데이터셋으로 그래프를 그리려면 다음 과정을 거치면 됩니다.

````python
# X_train 데이터를 사용해서 pandas.DataFrame을 만들고 iris_dataset.feature_names의 문자열을 사용해 그래프를 그립니다.
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# DataFrame을 사용해 y_train에 따라 색으로 구분된 산점도 행렬을 그립니다.
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o',
                           hist_kwds={'bins': 20}, s=60, alpha=8, cmap=mglearn.cm3)
````

- 세 클래스가 측정값에 따라 비교적 잘 구분되는 것을 토대로 머신러닝 모델을 학습할 수 있을 것입니다.

### 1.7.4 첫 번째 머신러닝 모델: k-최근접 이웃 알고리즘

- scikit-learn은 다양한 분류 알고리즘을 제공하고 있습니다만 여기서는 비교적 이해하기 쉬은 **k-최근접 이웃** 분류기를 사용합니다.

- 이 모델은 단순히 훈련 데이터를 저장해 만드는 것으로 새로운 데이터 포인트에 대한 예측을 가장 가까운 훈련 데이터 포인트를 통해 해답을 찾아갑니다.

    - 여기서 이름에서 보듯이 이 알고리즘은 가장 가까운 이웃 하나가 아닌 k개의 이웃을 찾아갑니다.
    
    - 그런 다음 가장 빈도가 높은 클래스를 예측값으로 사용합니다.

- 분류 방법은 다음과 같습니다.

````python
from sklearn.neighbors import KNeighborsClassifier  # 구현된 모델의 위치입니다.

# 가장 가까운 이웃을 하나만 찾겠습니다.
knn = KNeighborsClassifier(n_neighbors=1)

# knn은 훈련 데이터를 저장하고 있으며 모델을 만들기 위해 `fit` 메서드를 사용합니다.
knn.fit(X_train, y_train)
````

- `fit` 메서드는 knn 객체 자체를 반환하고 변경시킵니다.

    - 이 출력에서 knn에서 사용가능한 매개변수를 알아볼 수 있습니다.

### 1.7.5 예측하기

- 예측하기 위해 꽃받침의 길이가 5cm, 폭이 2.9cm, 꽃잎의 길이가 1cm, 폭이 0.2cm인 붓꽃을 넣고 품종을 확인해봅시다.

````python
# 예측용 데이터입니다.
X_new = np.array([[5, 2.9, 1, 0.2]])
print(f'X_new.shape: {X_new.shape}')

# knn 객체의 `predict` 메서드를 사용해 예측해봅시다.
prediction = knn.predict(X_new)
print(f'예측: {prediction[0]}')
print(f'예측한 이름: {iris_dataset["target_names"][prediction][0]}')
````

- 그렇다면 이 결과를 신뢰하는 방법을 알아봅시다.

### 1.7.6 모델 평가하기

- 앞서 만든 테스트 세트를 사용해 얼마나 많은 붓꽃의 품종을 맞췄는지 **정확도**를 확인할 수 있습니다.

    - 테스트 세트는 모델 생성에 사용되지 않았고 모든 붓꽃의 품종을 정확히 알고 있으므로 매우 유용합니다.

````python
# 테스트 세트를 사용해 예측합니다.
y_pred = knn.predict(X_test)
print(f'테스트 세트에 대한 예측값: {y_pred}')

# 정확도는 아래 두 가지 방법으로 계산할 수 있습니다.
print(f'테스트 세트의 정확도: {np.mean(y_pred == y_test):.2f}')
print(f'테스트 세트의 정확도: {knn.score(X_test, y_test):.2f}')
````

- 이 모델의 정확도는 약 `0.97`로 97%의 붓꽃의 품종을 맞출 수 있었습니다.

- 이 예제를 바탕으로 앞으로 모델의 성능을 높이는 방법과 모델을 튜닝할 때 주의할 점을 살펴보겠습니다.

## 1.8 요약 및 정리

- 처음엔 머신러닝과 머신러닝 애플리케이션에 대한 간단한 소개, 비지도 학습과 지도 학습의 차이 등을 확인하고 책에서 사용할 도구를 둘러보았습니다.

    - 실측한 자료를 바탕으로 붓꽃의 품종이 무엇인지 예측하는 작업을 진행했는데 이 작업은 정확한 품종으로 구분한 데이터셋을 사용했으므로 **지도 학습**입니다.
    
    - 또한, 품종이 정해져 있으므로 **분류** 문제입니다.
    
- 분류 문제에서 각 품종을 **클래스**라 하며 개별 붓꽃의 품종은 **레이블**이라고 합니다.

- 데이터셋은 일반적으로 NumPy 배열이며 데이터를 담은 하나를 scikit-learn에서는 X로, 정확한 혹은 기대하는 출력을 y로 표기합니다.

- 이 데이터셋을 모델 구축에 사용할 **훈련 세트**와 모델이 새로운 데이터에 얼마나 잘 적용될 수 있는지 평가하는 **테스트 세트**로 나눕니다.

- 앞서서 사용한 **k-최근접 이웃** 분류 알고리즘은 새 데이터 포인트를 예측하기 위해 훈련 데이터에서 가장 가까운 이웃을 선택합니다.

    - 이 알고리즘은 `KNeighborsClassifier` 클래스에 구현되어 있으며 `n_neighbors` 매개변수를 통해 클래스 객체를 만듭니다.
    
    - 그리고 훈련 데이터 `X_train`과 훈련 데이터의 레이블 `y_train`을 매개변수로 `fit`을 통해 모델을 구축합니다.
    
    - 그런 다음 모델의 정확도를 계산하는 `score` 메서드롤 모델을 평가합니다.
    
    - 이에 따라 나온 결과는 새로운 데이터에 대해 이 모델을 적용했을 때 해당 결과 비율로 신뢰할 수 있음을 알려줍니다.

- 이를 종합한 훈련과 평과 과정의 코드는 다음과 같습니다.

````python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

print(f'테스트 세트의 정확도: {knn.score(X_test, y_test):.2f}')
````

- 이 코드에는 scikit-learn 머신러닝 알고리즘들이 가진 핵심이 담겨 있습니다.

    - fit, predict, score 메서드는 scikit-learn 지도 학습 모델의 공통 인터페이스입니다.

- 이 장에서 소개한 개념과 함ㄲ 이런 모델들을 많은 머신러닝 작업에 적용할 수 있습니다.
